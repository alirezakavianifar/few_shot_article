```python
%%writefile /content/attMPTI/dataloaders/s3dis.py
""" Data Preprocess and Loader for S3DIS Dataset

Author: Zhao Na, 2020
"""
import os
import glob
import numpy as np
import pickle


class S3DISDataset(object):
    def __init__(self, cvfold, data_path):
        self.data_path = data_path
        self.classes = 13
        # self.class2type = {0:'ceiling', 1:'floor', 2:'wall', 3:'beam', 4:'column', 5:'window', 6:'door', 7:'table',
        #                    8:'chair', 9:'sofa', 10:'bookcase', 11:'board', 12:'clutter'}
        class_names = open(os.path.join(os.path.dirname(data_path), 'meta', 's3dis_classnames.txt')).readlines()
        self.class2type = {i: name.strip() for i, name in enumerate(class_names)}
        print(self.class2type)
        self.type2class = {self.class2type[t]: t for t in self.class2type}
        self.types = self.type2class.keys()
        self.fold_0 = ['beam', 'board', 'bookcase', 'ceiling', 'chair', 'column']
        self.fold_1 = ['door', 'floor', 'sofa', 'table', 'wall', 'window']

        if cvfold == 0:
            self.test_classes = [self.type2class[i] for i in self.fold_0]
        elif cvfold == 1:
            self.test_classes = [self.type2class[i] for i in self.fold_1]
        else:
            raise NotImplementedError('Unknown cvfold (%s). [Options: 0,1]' %cvfold)

        all_classes = [i for i in range(0, self.classes-1)]
        self.train_classes = [c for c in all_classes if c not in self.test_classes]

        # print('train_class:{0}'.format(self.train_classes))
        # print('test_class:{0}'.format(self.test_classes))

        self.class2scans = self.get_class2scans()

    def get_class2scans(self):
        class2scans_file = os.path.join(self.data_path, 'class2scans.pkl')
        if os.path.exists(class2scans_file):
            #load class2scans (dictionary)
            with open(class2scans_file, 'rb') as f:
                class2scans = pickle.load(f)
        else:
            min_ratio = .05  # to filter out scans with only rare labelled points
            min_pts = 100  # to filter out scans with only rare labelled points
            class2scans = {k:[] for k in range(self.classes)}

            for file in glob.glob(os.path.join(self.data_path, 'data', '*.npy')):
                scan_name = os.path.basename(file)[:-4]
                data = np.load(file)
                labels = data[:,6].astype(int)
                classes = np.unique(labels)
                print('{0} | shape: {1} | classes: {2}'.format(scan_name, data.shape, list(classes)))
                for class_id in classes:
                    #if the number of points for the target class is too few, do not add this sample into the dictionary
                    num_points = np.count_nonzero(labels == class_id)
                    threshold = max(int(data.shape[0]*min_ratio), min_pts)
                    if num_points > threshold:
                        class2scans[class_id].append(scan_name)

            print('==== class to scans mapping is done ====')
            for class_id in range(self.classes):
                print('\t class_id: {0} | min_ratio: {1} | min_pts: {2} | class_name: {3} | num of scans: {4}'.format(
                          class_id,  min_ratio, min_pts, self.class2type[class_id], len(class2scans[class_id])))

            with open(class2scans_file, 'wb') as f:
                pickle.dump(class2scans, f, pickle.HIGHEST_PROTOCOL)
        return class2scans
```

```python
%%writefile /content/attMPTI/models/adapter.py

import torch
import torch.nn as nn

class DistributionShiftAdapter(nn.Module):
    def __init__(self, dim):
        super().__init__()
        self.mu_adapter = nn.Sequential(
            nn.Linear(dim, dim),
            nn.ReLU(),
            nn.Linear(dim, dim)
        )
        self.cov_adapter = nn.Sequential(
            nn.Linear(dim, dim),
            nn.ReLU(),
            nn.Linear(dim, dim)
        )

    def forward(self, mu_s, Sigma_s):
        """
        Args:
            mu_s: [D] or [B, D]
            Sigma_s: [D, D] or [B, D, D]
        Returns:
            mu_adj: same shape as mu_s
            Sigma_adj: same shape as Sigma_s
        """
        was_unbatched = False
        if mu_s.dim() == 1:
            was_unbatched = True
            mu_s = mu_s.unsqueeze(0)      # [1, D]
            Sigma_s = Sigma_s.unsqueeze(0)  # [1, D, D]

        delta_mu = self.mu_adapter(mu_s)  # [B, D]
        diag_cov = torch.diagonal(Sigma_s, dim1=-2, dim2=-1)  # [B, D]
        delta_cov = self.cov_adapter(diag_cov)  # [B, D]
        mu_adj = mu_s + delta_mu
        Sigma_adj = Sigma_s + torch.diag_embed(delta_cov)

        if was_unbatched:
            mu_adj = mu_adj.squeeze(0)
            Sigma_adj = Sigma_adj.squeeze(0)

        return mu_adj, Sigma_adj
```

```python
%%writefile /content/attMPTI/models/adapter.py

import torch
import torch.nn as nn

class DistributionShiftAdapter(nn.Module):
    def __init__(self, dim):
        super().__init__()
        self.adapter = nn.Sequential(
            nn.Linear(dim, dim),
            nn.ReLU(),
            nn.Linear(dim, dim)
        )

    def forward(self, mu_s):
        """
        Args:
            mu_s: [D] or [B, D] — assumed to be L2-normalized or raw mean vector
        Returns:
            mu_adj: same shape as mu_s, L2-normalized
        """
        was_unbatched = False
        if mu_s.dim() == 1:
            was_unbatched = True
            mu_s = mu_s.unsqueeze(0)  # [1, D]

        # Ensure input is normalized (robustness)
        mu_s_norm = mu_s / (mu_s.norm(dim=-1, keepdim=True) + 1e-8)

        # Predict residual in tangent space (approx.)
        delta = self.adapter(mu_s_norm)

        # Add residual and renormalize
        mu_adj = mu_s_norm + delta
        mu_adj = mu_adj / (mu_adj.norm(dim=-1, keepdim=True) + 1e-8)

        if was_unbatched:
            mu_adj = mu_adj.squeeze(0)

        return mu_adj
```

```python
%%writefile /content/attMPTI/models/bayesian_classifier.py

import torch
import torch.nn as nn
import torch.nn.functional as F

class BayesianClassifier(nn.Module):
    def __init__(self, num_classes, embedding_dim, beta=8.0, precision=1.0):
        super(BayesianClassifier, self).__init__()
        self.num_classes = num_classes
        self.embedding_dim = embedding_dim
        self.beta = beta
        self.precision_scalar = precision

        self.register_buffer('mean', torch.zeros(num_classes, embedding_dim))
        self.register_buffer('var', torch.ones(num_classes, embedding_dim))

        self.register_buffer('prior_mean', torch.zeros(num_classes, embedding_dim))
        self.register_buffer('prior_var', torch.ones(num_classes, embedding_dim))

    @staticmethod
    def construct_prior_for_novel_class(z_support, base_priors=None, beta=8.0, mode='hybrid', lambda_mix=0.5):
        device = z_support.device
        if z_support.dim() == 2:
            z_centroid = z_support.mean(dim=0)
        else:
            z_centroid = z_support  # [D]
        D = z_centroid.size(0)

        # Step 1: Base-driven prior (only mean and var)
        if base_priors is not None and len(base_priors) > 0:
            base_items = list(base_priors.items())
            base_mus = torch.stack([mu for _, (mu, var_or_cov) in base_items], dim=0)  # [B, D]

            first_var_or_cov = base_items[0][1][1]
            if first_var_or_cov.dim() == 1:
                base_vars = torch.stack([var_or_cov for _, (_, var_or_cov) in base_items], dim=0)
            else:
                base_vars = torch.stack([torch.diagonal(var_or_cov) for _, (_, var_or_cov) in base_items], dim=0)

            zc = z_centroid / (z_centroid.norm() + 1e-8)
            bm = base_mus / (base_mus.norm(dim=1, keepdim=True) + 1e-8)
            sims = F.cosine_similarity(zc.unsqueeze(0), bm, dim=1)
            weights = torch.softmax(beta * sims, dim=0)

            mu_base = torch.sum(weights.unsqueeze(1) * base_mus, dim=0)
            var_base = torch.sum(weights.unsqueeze(1) * base_vars, dim=0)
        else:
            mu_base = torch.zeros(D, device=device)
            var_base = torch.ones(D, device=device)

        # Step 2: Self prior
        mu_self = z_support.mean(dim=0)
        N = z_support.size(0)
        if N <= 1:
            var_self = torch.ones(D, device=device) * 1e-6
        else:
            var_self = z_support.var(dim=0, unbiased=False) + 1e-6

        # Step 3: Hybrid
        if mode == 'base':
            mu_prior = mu_base
            var_prior = var_base
        elif mode == 'self':
            mu_prior = mu_self
            var_prior = var_self
        elif mode == 'hybrid':
            mu_prior = (1 - lambda_mix) * mu_self + lambda_mix * mu_base
            var_prior = (1 - lambda_mix) * var_self + lambda_mix * var_base
        else:
            raise ValueError(f"Unknown mode '{mode}'. Choose from ['base','self','hybrid'].")

        return mu_prior, var_prior

    def update_posterior(self, support_set_features, support_set_labels, base_priors=None, adapter=None):
        device = support_set_features.device
        D = self.embedding_dim

        for c in range(self.num_classes):
            mask = (support_set_labels == c)
            if mask.sum() == 0:
                continue

            support_set = support_set_features[mask]
            Nc = support_set.size(0)

            if base_priors is not None:
                mu_hybrid, var_hybrid = self.construct_prior_for_novel_class(
                    support_set, base_priors, beta=self.beta, mode='hybrid', lambda_mix=0.4
                )
            else:
                mu_hybrid = support_set.mean(dim=0)
                if Nc <= 1:
                    var_hybrid = torch.ones(D, device=device) * 1e-6
                else:
                    var_hybrid = support_set.var(dim=0, unbiased=False) + 1e-6

            if adapter is not None:
                with torch.no_grad():
                    Sigma_hybrid = torch.diag(var_hybrid)
                    mu_prior, Sigma_prior = adapter(mu_hybrid, Sigma_hybrid)
                    var_prior = torch.diagonal(Sigma_prior)
            else:
                mu_prior = mu_hybrid
                var_prior = var_hybrid

            support_mean = support_set.mean(dim=0)
            prec_prior = 1.0 / var_prior
            prec_likelihood = self.precision_scalar

            posterior_prec = prec_prior + Nc * prec_likelihood
            posterior_mean = (prec_prior * mu_prior + prec_likelihood * Nc * support_mean) / posterior_prec
            posterior_var = 1.0 / posterior_prec

            # Ensure no NaN or Inf
            posterior_mean = torch.nan_to_num(posterior_mean, nan=0.0, posinf=1e6, neginf=-1e6)
            posterior_var = torch.clamp(posterior_var, min=1e-8, max=1e6)

            self.mean[c] = posterior_mean
            self.var[c] = posterior_var

    def predict(self, z_q_batch):
        mean_norm = self.mean / (self.mean.norm(dim=1, keepdim=True) + 1e-8)
        z_norm = z_q_batch / (z_q_batch.norm(dim=1, keepdim=True) + 1e-8)
        logits = torch.mm(z_norm, mean_norm.t())
        return logits
```

```python
%%writefile /content/attMPTI/models/bayesian_classifier.py

import torch
import torch.nn as nn
import torch.nn.functional as F

class BayesianClassifier(nn.Module):
    def __init__(self, num_classes, embedding_dim, beta=8.0, precision=1.0, temperature=1.0):
        super(BayesianClassifier, self).__init__()
        self.num_classes = num_classes
        self.embedding_dim = embedding_dim
        self.beta = beta
        self.precision_scalar = precision
        self.temperature = temperature  # for scaling logits

        self.register_buffer('mean', torch.zeros(num_classes, embedding_dim))
        self.register_buffer('var', torch.ones(num_classes, embedding_dim))

        self.register_buffer('prior_mean', torch.zeros(num_classes, embedding_dim))
        self.register_buffer('prior_var', torch.ones(num_classes, embedding_dim))

    @staticmethod
    def construct_prior_for_novel_class(z_support, base_priors=None, beta=8.0, mode='hybrid', lambda_mix=0.5):
        device = z_support.device
        if z_support.dim() == 2:
            z_centroid = z_support.mean(dim=0)
        else:
            z_centroid = z_support  # [D]
        D = z_centroid.size(0)

        # Step 1: Base-driven prior (only mean and var)
        if base_priors is not None and len(base_priors) > 0:
            base_items = list(base_priors.items())
            base_mus = torch.stack([mu for _, (mu, var_or_cov) in base_items], dim=0)  # [B, D]

            first_var_or_cov = base_items[0][1][1]
            if first_var_or_cov.dim() == 1:
                base_vars = torch.stack([var_or_cov for _, (_, var_or_cov) in base_items], dim=0)
            else:
                base_vars = torch.stack([torch.diagonal(var_or_cov) for _, (_, var_or_cov) in base_items], dim=0)

            zc = z_centroid / (z_centroid.norm() + 1e-8)
            bm = base_mus / (base_mus.norm(dim=1, keepdim=True) + 1e-8)
            sims = F.cosine_similarity(zc.unsqueeze(0), bm, dim=1)
            weights = torch.softmax(beta * sims, dim=0)

            mu_base = torch.sum(weights.unsqueeze(1) * base_mus, dim=0)
            var_base = torch.sum(weights.unsqueeze(1) * base_vars, dim=0)
        else:
            mu_base = torch.zeros(D, device=device)
            var_base = torch.ones(D, device=device)

        # Step 2: Self prior
        mu_self = z_support.mean(dim=0)
        N = z_support.size(0)
        if N <= 1:
            var_self = torch.ones(D, device=device) * 1e-6
        else:
            var_self = z_support.var(dim=0, unbiased=False) + 1e-6

        # Step 3: Hybrid
        if mode == 'base':
            mu_prior = mu_base
            var_prior = var_base
        elif mode == 'self':
            mu_prior = mu_self
            var_prior = var_self
        elif mode == 'hybrid':
            mu_prior = (1 - lambda_mix) * mu_self + lambda_mix * mu_base
            var_prior = (1 - lambda_mix) * var_self + lambda_mix * var_base
        else:
            raise ValueError(f"Unknown mode '{mode}'. Choose from ['base','self','hybrid'].")

        return mu_prior, var_prior

    def update_posterior(self, support_set_features, support_set_labels, base_priors=None, adapter=None):
        device = support_set_features.device
        D = self.embedding_dim

        for c in range(self.num_classes):
            mask = (support_set_labels == c)
            if mask.sum() == 0:
                continue

            support_set = support_set_features[mask]
            Nc = support_set.size(0)

            if base_priors is not None:
                mu_hybrid, var_hybrid = self.construct_prior_for_novel_class(
                    support_set, base_priors, beta=self.beta, mode='hybrid', lambda_mix=0.4
                )
            else:
                mu_hybrid = support_set.mean(dim=0)
                if Nc <= 1:
                    var_hybrid = torch.ones(D, device=device) * 1e-6
                else:
                    var_hybrid = support_set.var(dim=0, unbiased=False) + 1e-6

            if adapter is not None:
                with torch.no_grad():
                    Sigma_hybrid = torch.diag(var_hybrid)
                    mu_prior, Sigma_prior = adapter(mu_hybrid, Sigma_hybrid)
                    var_prior = torch.diagonal(Sigma_prior)
            else:
                mu_prior = mu_hybrid
                var_prior = var_hybrid

            support_mean = support_set.mean(dim=0)
            prec_prior = 1.0 / (var_prior + 1e-12)  # for stability
            prec_likelihood = self.precision_scalar

            posterior_prec = prec_prior + Nc * prec_likelihood
            posterior_mean = (prec_prior * mu_prior + prec_likelihood * Nc * support_mean) / posterior_prec
            posterior_var = 1.0 / (posterior_prec + 1e-12)

            # Ensure no NaN or Inf
            posterior_mean = torch.nan_to_num(posterior_mean, nan=0.0, posinf=1e6, neginf=-1e6)
            posterior_var = torch.clamp(posterior_var, min=1e-8, max=1e6)

            self.mean[c] = posterior_mean.detach()
            self.var[c] = posterior_var.detach()

    def predict(self, z_q_batch):
        """
        Improved prediction using precision-weighted feature alignment.
        z_q_batch: [Q, D]
        Returns: [Q, C] logits
        """
        # Avoid in-place modification; use buffers safely
        mean = self.mean  # [C, D]
        var = self.var    # [C, D]

        # Compute precision (inverse variance)
        prec = 1.0 / (var + 1e-8)  # [C, D]

        # Weight features by sqrt(precision) for Mahalanobis-like alignment
        mean_weighted = mean * torch.sqrt(prec)  # [C, D]
        z_weighted = z_q_batch.unsqueeze(1) * torch.sqrt(prec).unsqueeze(0)  # [Q, C, D]

        # Weighted dot product (numerator of cosine)
        weighted_dot = (z_weighted * mean_weighted.unsqueeze(0)).sum(dim=2)  # [Q, C]

        # Compute norms for denominator
        z_norm = torch.norm(z_weighted, dim=2)  # [Q, C]
        mean_norm = torch.norm(mean_weighted, dim=1).unsqueeze(0)  # [1, C]

        # Cosine similarity with precision weighting
        cos_sim = weighted_dot / (z_norm * mean_norm + 1e-8)  # [Q, C]

        # Apply temperature for smoother logits (helps IoU by reducing overconfident errors)
        logits = cos_sim / self.temperature

        return logits
```

```python
%%writefile /content/attMPTI/models/bayesian_classifier.py

import torch
import torch.nn as nn
import torch.nn.functional as F

class BayesianClassifier(nn.Module):
    def __init__(self, num_classes, embedding_dim, beta=8.0, temperature=1.0):
        super(BayesianClassifier, self).__init__()
        self.num_classes = num_classes
        self.embedding_dim = embedding_dim
        self.beta = beta
        self.temperature = temperature

        # Only store normalized class means (no variances!)
        self.register_buffer('mean', torch.zeros(num_classes, embedding_dim))

    @staticmethod
    def construct_prior_for_novel_class(z_support, base_priors=None, beta=8.0, lambda_mix=0.5):
        device = z_support.device
        z_self = z_support.mean(dim=0)
        z_self = z_self / (z_self.norm() + 1e-8)

        if base_priors is not None and len(base_priors) > 0:
            # ✅ حالا base_priors فقط mu دارد، نه (mu, var)
            base_mus = torch.stack(list(base_priors.values()), dim=0)  # [B, D]
            base_mus_norm = base_mus / (base_mus.norm(dim=1, keepdim=True) + 1e-8)

            zc = z_self.unsqueeze(0)  # [1, D]
            sims = F.cosine_similarity(zc, base_mus_norm, dim=1)  # [B]
            weights = torch.softmax(beta * sims, dim=0)

            mu_base = torch.sum(weights.unsqueeze(1) * base_mus_norm, dim=0)
            mu_base = mu_base / (mu_base.norm() + 1e-8)
        else:
            mu_base = torch.zeros_like(z_self)
            mu_base = mu_base / (mu_base.norm() + 1e-8)

        # Interpolate in normalized space
        mu_prior = (1 - lambda_mix) * z_self + lambda_mix * mu_base
        mu_prior = mu_prior / (mu_prior.norm() + 1e-8)
        return mu_prior

    def update_posterior(self, support_set_features, support_set_labels, base_priors=None, adapter=None):
        """
        Update class means as normalized hybrid prototypes.
        No variances. Fully compatible with cosine similarity.
        """
        device = support_set_features.device
        for c in range(self.num_classes):
            mask = (support_set_labels == c)
            if mask.sum() == 0:
                continue

            support_set = support_set_features[mask]  # [M, D]
            mu_self = support_set.mean(dim=0)
            mu_self = mu_self / (mu_self.norm() + 1e-8)

            if base_priors is not None:
                mu_prior = self.construct_prior_for_novel_class(
                    support_set, base_priors, beta=self.beta, lambda_mix=0.4
                )
            else:
                mu_prior = mu_self

            # Optional: apply adapter (now only on mu, and differentiable!)
            if adapter is not None:
                mu_prior = adapter(mu_prior)  # already normalized inside adapter

            # Final posterior = prior (already hybrid) — or you can blend with mu_self again
            posterior_mean = mu_prior  # simplest and most stable

            # Store normalized mean
            self.mean[c] = posterior_mean.detach()

    def predict(self, z_q_batch):
        """
        Pure cosine similarity on normalized features and class means.
        """
        # Normalize query features
        z_norm = z_q_batch / (z_q_batch.norm(dim=1, keepdim=True) + 1e-8)
        mean_norm = self.mean / (self.mean.norm(dim=1, keepdim=True) + 1e-8)

        # Cosine similarity
        cos_sim = torch.mm(z_norm, mean_norm.t())  # [Q, C]

        # Apply temperature for calibration
        logits = cos_sim / self.temperature
        return logits
```

```python
%%writefile /content/attMPTI/runs/mpti_train.py

""" Attention-aware Multi-Prototype Transductive Inference
    — Simplified Bayesian + Differentiable Adapter (Normalized Mean-Only)
"""
import os
import torch
from torch.utils.data import DataLoader
from torch.utils.tensorboard import SummaryWriter

from runs.eval import test_few_shot
from dataloaders.loader import MyDataset, MyTestDataset, MyFullDataset, batch_test_task_collate
from models.mpti_learner import MPTILearner
from utils.cuda_util import cast_cuda
from utils.logger import init_logger


def collect_base_priors(model, dataloader, num_base_classes, feat_dim, device):
    model.eval()
    all_feats = [[] for _ in range(num_base_classes)]

    with torch.no_grad():
        for points, labels in dataloader:
            points = points.to(device)
            labels = labels.to(device)
            points = points.transpose(1, 2).contiguous()
            feats = model.getFeatures(points)
            feats = feats.transpose(1, 2).contiguous().view(-1, feat_dim)
            labels_flat = labels.view(-1)

            for c in range(num_base_classes):
                mask = (labels_flat == c)
                if mask.any():
                    class_feats = feats[mask]
                    all_feats[c].append(class_feats)

    base_priors = {}
    for c in range(num_base_classes):
        if len(all_feats[c]) > 0:
            feats_c = torch.cat(all_feats[c], dim=0)
            mu = feats_c.mean(dim=0)
            mu = mu / (mu.norm() + 1e-8)
            base_priors[c] = mu
        else:
            base_priors[c] = torch.zeros(feat_dim, device=device)

    return base_priors


def train(args):
    logger = init_logger(args.log_dir, args)
    MPTI = MPTILearner(args)
    model = MPTI.model
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

    # Adapter optimizer (now trained end-to-end with main loss)
    adapter_optimizer = None
    if getattr(args, 'use_adapter', False):
        adapter_params = list(model.distribution_adapter.parameters())
        adapter_optimizer = torch.optim.Adam(adapter_params, lr=args.lr * 0.1)

    # === Collect base priors (mean-only, normalized) ===
    if getattr(args, 'use_bayesian', False):
        logger.cprint("=> Collecting base priors (normalized means)...")
        BASE_DATASET = MyFullDataset(
            args.data_path, args.dataset, cvfold=args.cvfold,
            phase='train', classes=[1,2,5,6,7,9],
            num_point=args.pc_npts, pc_attribs=args.pc_attribs, pc_augm=False
        )
        BASE_LOADER = DataLoader(BASE_DATASET, batch_size=8, shuffle=False, num_workers=0)
        model.base_priors = collect_base_priors(model, BASE_LOADER, num_base_classes=6, feat_dim=model.feat_dim, device=device)
        logger.cprint(f"=> Collected normalized base priors for {len(model.base_priors)} classes.")

    # === Episodic datasets ===
    PC_AUGMENT_CONFIG = {
        'scale': args.pc_augm_scale,
        'rot': args.pc_augm_rot,
        'mirror_prob': args.pc_augm_mirror_prob,
        'jitter': args.pc_augm_jitter
    }

    TRAIN_DATASET = MyDataset(
        args.data_path, args.dataset, cvfold=args.cvfold, num_episode=args.n_iters,
        n_way=args.n_way, k_shot=args.k_shot, n_queries=args.n_queries,
        phase=args.phase, mode='train',
        num_point=args.pc_npts, pc_attribs=args.pc_attribs,
        pc_augm=args.pc_augm, pc_augm_config=PC_AUGMENT_CONFIG
    )

    VALID_DATASET = MyTestDataset(
        args.data_path, args.dataset, cvfold=args.cvfold,
        num_episode_per_comb=args.n_episode_test,
        n_way=args.n_way, k_shot=args.k_shot, n_queries=args.n_queries,
        num_point=args.pc_npts, pc_attribs=args.pc_attribs
    )
    VALID_CLASSES = list(VALID_DATASET.classes)

    TRAIN_LOADER = DataLoader(TRAIN_DATASET, batch_size=1, collate_fn=batch_test_task_collate, num_workers=0)
    VALID_LOADER = DataLoader(VALID_DATASET, batch_size=1, collate_fn=None, num_workers=0)

    WRITER = SummaryWriter(log_dir=args.log_dir)
    best_iou = 0

    for batch_idx, data in enumerate(TRAIN_LOADER):
        if torch.cuda.is_available():
            data = tuple(x.cuda(non_blocking=True) for x in data)
        support_x, support_y, query_x, query_y = data
        model.train()

        if getattr(args, 'use_bayesian', False):
            query_logits, loss_main = model._forward_bayesian(support_x, support_y, query_x, query_y)
        else:
            query_logits, loss_main = model._forward_label_propagation(support_x, support_y, query_x, query_y)

        query_pred = query_logits.argmax(dim=1)
        correct = torch.eq(query_pred, query_y).sum().item()
        accuracy = correct / query_y.numel()

        # 📌 Optimize main model + adapter together
        MPTI.optimizer.zero_grad()
        if adapter_optimizer is not None:
            adapter_optimizer.zero_grad()

        loss_main.backward()
        MPTI.optimizer.step()
        if adapter_optimizer is not None:
            adapter_optimizer.step()  # ✅ Now trained end-to-end
        MPTI.lr_scheduler.step()

        # 📊 Logging
        logger.cprint('==[Train] Iter: %d | Loss: %.4f | Acc: %.4f ==' %
                      (batch_idx, loss_main.item(), accuracy))
        WRITER.add_scalar('Train/loss', loss_main.item(), batch_idx)
        WRITER.add_scalar('Train/accuracy', accuracy, batch_idx)

        # 🔍 Validation
        if (batch_idx + 1) % args.eval_interval == 0:
            valid_loss, mean_IoU = test_few_shot(VALID_LOADER, MPTI, logger, VALID_CLASSES)
            logger.cprint('\n=====[VALID] Loss: %.4f | Mean IoU: %.4f  =====\n' % (valid_loss, mean_IoU))
            WRITER.add_scalar('Valid/loss', valid_loss, batch_idx)
            WRITER.add_scalar('Valid/meanIoU', mean_IoU, batch_idx)

            if mean_IoU > best_iou:
                best_iou = mean_IoU
                logger.cprint('*******************Model Saved*******************')
                save_dict = {
                    'iteration': batch_idx + 1,
                    'model_state_dict': model.state_dict(),
                    'optimizer_state_dict': MPTI.optimizer.state_dict(),
                    'adapter_optimizer_state_dict': adapter_optimizer.state_dict() if adapter_optimizer else None,
                    'loss': valid_loss,
                    'IoU': best_iou
                }
                if hasattr(model, 'base_priors'):
                    saved_priors = {}
                    for k, mu in model.base_priors.items():
                        saved_priors[k] = mu.cpu()
                    save_dict['base_priors'] = saved_priors
                torch.save(save_dict, os.path.join(args.log_dir, 'checkpoint.tar'))

        if torch.cuda.is_available():
            torch.cuda.empty_cache()

    WRITER.close()
```

```python
%%writefile /content/attMPTI/models/mpti_learner.py

""" MPTI with/without attention Learner for Few-shot 3D Point Cloud Semantic Segmentation

Author: Zhao Na, 2020
Mod (ADMP + Bayesian + Adapter): 2025 — Optimized for Colab (lightweight + high accuracy)
"""
import os
import torch
from torch import optim
from torch.nn import functional as F

from models.mpti import MultiPrototypeTransductiveInference
from utils.checkpoint_util import load_pretrain_checkpoint, load_model_checkpoint


class MPTILearner(object):
    def __init__(self, args, mode='train'):
        self.args = args
        self.train_stage = getattr(args, 'train_stage', None)
        self.use_bayesian = getattr(args, 'use_bayesian', False)
        self.use_adapter = getattr(args, 'use_adapter', False)

        # ADMP policy options (unchanged)
        self.use_depth_policy = getattr(args, 'use_depth_policy', False)
        self.policy_bins = int(getattr(args, 'policy_bins', 3))
        self.policy_depth_for_graph = int(getattr(args, 'policy_depth_for_graph', 1))
        self.policy_select_via = getattr(args, 'policy_select_via', 'rule')
        self.policy_val_episodes = int(getattr(args, 'policy_val_episodes', 0))

        # init model
        self.model = MultiPrototypeTransductiveInference(args)
        device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
        self.model = self.model.to(device)

        if mode == 'train':
            if self.train_stage is not None:
                self._configure_stage(self.train_stage)

            # Build main optimizer
            self._build_optimizer_and_scheduler(args)

            # Build adapter optimizer if needed
            if self.use_adapter:
                adapter_params = list(self.model.distribution_adapter.parameters())
                self.adapter_optimizer = torch.optim.Adam(adapter_params, lr=args.lr * 0.1)
            else:
                self.adapter_optimizer = None

            # Load checkpoint
            if args.model_checkpoint_path is not None:
                ckpt = torch.load(args.model_checkpoint_path, map_location=device)
                self.model.load_state_dict(ckpt['model_state_dict'])
                self.optimizer.load_state_dict(ckpt['optimizer_state_dict'])
                if self.adapter_optimizer and 'adapter_optimizer_state_dict' in ckpt:
                    self.adapter_optimizer.load_state_dict(ckpt['adapter_optimizer_state_dict'])
                # Load base_priors (now stored as (mean, var))
                if 'base_priors' in ckpt:
                    device = next(self.model.parameters()).device
                    loaded_priors = {}
                    for k, (mu, var) in ckpt['base_priors'].items():
                        loaded_priors[k] = (mu.to(device), var.to(device))
                    self.model.base_priors = loaded_priors

        elif mode == 'test':
            device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
            ckpt = torch.load(args.model_checkpoint_path, map_location=device)
            self.model.load_state_dict(ckpt['model_state_dict'])
            if 'base_priors' in ckpt:
                loaded_priors = {}
                for k, (mu, var) in ckpt['base_priors'].items():
                    loaded_priors[k] = (mu.to(device), var.to(device))
                self.model.base_priors = loaded_priors
        else:
            raise ValueError('Wrong MPTILearner mode (%s)! Option: train/test' % mode)

    def _set_requires_grad(self, module, flag: bool):
        if module is None:
            return
        for p in module.parameters():
            p.requires_grad = flag

    def _configure_stage(self, stage: int):
        pass

    def _build_optimizer_and_scheduler(self, args):
        adapter_param_set = set()
        if self.use_adapter:
            adapter_param_set = set(self.model.distribution_adapter.parameters())

        main_params = []
        for name, p in self.model.named_parameters():
            if p not in adapter_param_set:
                p.requires_grad = True
                main_params.append(p)

        if len(main_params) == 0:
            raise ValueError("❌ No main trainable parameters found!")

        self.optimizer = torch.optim.Adam(main_params, lr=args.lr)
        self.lr_scheduler = optim.lr_scheduler.CyclicLR(
            self.optimizer, args.lr, 0.0025, step_size_up=10,
            step_size_down=args.step_size, mode='exp_range', gamma=args.gamma,
            cycle_momentum=False, base_momentum=0.8, max_momentum=0.9
        )

    def train(self, data):
        [support_x, support_y, query_x, query_y] = data
        self.model.train()
        device = support_x.device

        # ✅ استفاده از مسیر آموزشی پارامتری (نه label propagation!)
        query_logits, loss_main = self.model._forward_train(query_x, query_y)

        # Adapter training (if enabled) — optional, can be disabled in train
        loss_adapter = torch.tensor(0.0, device=device)
        if self.use_adapter:
            with torch.no_grad():
                sup_feat = self.model.getFeatures(support_x.view(-1, self.args.pc_in_dim, self.args.pc_npts))
                qry_feat = self.model.getFeatures(query_x.view(-1, self.args.pc_in_dim, self.args.pc_npts))

            sup_flat = sup_feat.transpose(1, 2).contiguous().view(-1, self.model.feat_dim)
            qry_flat = qry_feat.transpose(1, 2).contiguous().view(-1, self.model.feat_dim)

            mu_s = sup_flat.mean(dim=0)
            var_s = sup_flat.var(dim=0, unbiased=True) + 1e-6
            mu_q = qry_flat.mean(dim=0)
            var_q = qry_flat.var(dim=0, unbiased=True) + 1e-6

            Sigma_s = torch.diag(var_s)
            Sigma_q = torch.diag(var_q)

            mu_q_hat, Sigma_q_hat = self.model.distribution_adapter(mu_s, Sigma_s)
            var_q_hat = torch.diagonal(Sigma_q_hat)

            loss_mu = F.mse_loss(mu_q_hat, mu_q)
            loss_cov = F.mse_loss(var_q_hat, var_q)
            loss_adapter = loss_mu + loss_cov

            self.adapter_optimizer.zero_grad()
            loss_adapter.backward()
            self.adapter_optimizer.step()

        self.optimizer.zero_grad()
        loss_main.backward()
        self.optimizer.step()
        self.lr_scheduler.step()

        bs = query_x.shape[0]
        query_pred = query_logits.argmax(dim=1)  # [B, N]
        correct = torch.eq(query_pred, query_y).sum().item()
        accuracy = correct / query_y.numel()
        return loss_main + loss_adapter, accuracy

    def test(self, data):
        [support_x, support_y, query_x, query_y] = data
        self.model.eval()
        bs = support_x.shape[0]
        device = support_x.device

        with torch.no_grad():
            if self.use_bayesian:
                query_logits, loss = self.model._forward_bayesian(support_x, support_y, query_x, query_y)
            else:
                if (self.train_stage is None) and (not self.use_depth_policy):
                    query_logits, loss = self.model._forward_label_propagation(support_x, support_y, query_x, query_y)
                else:
                    raise NotImplementedError("ADMP + Bayesian not implemented together yet.")

            pred = F.softmax(query_logits, dim=1).argmax(dim=1)
            correct = torch.eq(pred, query_y.reshape(bs, -1)).sum().item()
            accuracy = correct / (query_y.numel())
            return pred, loss, accuracy
```

```python
%%writefile /content/attMPTI/models/mpti.py

""" Multi-prototype transductive inference with Bayesian adaptation for distribution shift

Author: Zhao Na, 2020 — Extended with Bayesian + Distribution Shift Adapter (Optimized for Colab + Trainable)
"""
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch_cluster import fps

from models.dgcnn import DGCNN
from models.attention import SelfAttention
from models.bayesian_classifier import BayesianClassifier
from models.adapter import DistributionShiftAdapter
from sklearn.neighbors import NearestNeighbors


class BaseLearner(nn.Module):
    """The class for inner loop."""
    def __init__(self, in_channels, params):
        super(BaseLearner, self).__init__()
        self.num_convs = len(params)
        self.convs = nn.ModuleList()
        for i in range(self.num_convs):
            in_dim = in_channels if i == 0 else params[i-1]
            self.convs.append(nn.Sequential(
                nn.Conv1d(in_dim, params[i], 1),
                nn.BatchNorm1d(params[i])
            ))

    def forward(self, x):
        for i in range(self.num_convs):
            x = self.convs[i](x)
            if i != self.num_convs - 1:
                x = F.relu(x)
        return x


class MultiPrototypeTransductiveInference(nn.Module):
    def __init__(self, args):
        super(MultiPrototypeTransductiveInference, self).__init__()
        self.n_way = args.n_way
        self.k_shot = args.k_shot
        self.in_channels = args.pc_in_dim
        self.n_points = args.pc_npts
        self.use_attention = args.use_attention
        self.n_subprototypes = args.n_subprototypes
        self.k_connect = args.k_connect
        self.sigma = args.sigma
        self.n_classes = self.n_way + 1

        # Backbone
        self.encoder = DGCNN(args.edgeconv_widths, args.dgcnn_mlp_widths, args.pc_in_dim, k=args.dgcnn_k)
        self.base_learner = BaseLearner(args.dgcnn_mlp_widths[-1], args.base_widths)

        if self.use_attention:
            self.att_learner = SelfAttention(args.dgcnn_mlp_widths[-1], args.output_dim)
        else:
            self.linear_mapper = nn.Conv1d(args.dgcnn_mlp_widths[-1], args.output_dim, 1, bias=False)

        raw_feat_dim = args.edgeconv_widths[0][-1] + args.output_dim + args.base_widths[-1]
        self.projected_dim = getattr(args, 'projected_dim', 128)  # default: 128
        self.feature_projector = nn.Linear(raw_feat_dim, self.projected_dim)
        self.feat_dim = self.projected_dim

        # ✅ Classifier for training (S3DIS has 13 semantic classes)
        self.train_classifier = nn.Linear(self.feat_dim, 13)

        # Bayesian components
        self.use_bayesian = getattr(args, 'use_bayesian', False)
        if self.use_bayesian:
            self.bayesian_classifier = BayesianClassifier(
                num_classes=self.n_classes,
                embedding_dim=self.feat_dim,
                beta=getattr(args, 'beta', 8.0),
                precision=getattr(args, 'precision', 1.0)
            )
            self.use_adapter = getattr(args, 'use_adapter', False)
            if self.use_adapter:
                self.distribution_adapter = DistributionShiftAdapter(self.feat_dim)

    def forward(self, support_x, support_y, query_x, query_y):
        if self.use_bayesian and not self.training:
            return self._forward_bayesian(support_x, support_y, query_x, query_y)
        elif not self.training:
            return self._forward_label_propagation(support_x, support_y, query_x, query_y)
        else:
            # ✅ Training mode: use parametric loss
            return self._forward_train(query_x, query_y)

    def _forward_train(self, query_x, query_y):
        """
        Parametric training path: classify each point in query.
        query_x: [B, C, N]
        query_y: [B, N] with class indices (0-12 for S3DIS)
        """
        B, C, N = query_x.shape
        query_x_flat = query_x.view(-1, C, N)  # [B, C, N]
        feats = self.getFeatures(query_x_flat)  # [B, D, N]
        feats = feats.transpose(1, 2).contiguous().view(-1, self.feat_dim)  # [B*N, D]

        logits = self.train_classifier(feats)  # [B*N, 13]
        labels = query_y.view(-1).long()       # [B*N]

        loss = F.cross_entropy(logits, labels)
        # Reshape logits to [B, N, 13] then [B, 13, N] for compatibility
        logits = logits.view(B, N, 13).transpose(1, 2).contiguous()
        return logits, loss

    def _forward_label_propagation(self, support_x, support_y, query_x, query_y,
                                   cached_support_feat=None, cached_query_feat=None):
        device = support_x.device

        if cached_support_feat is not None:
            support_feat = cached_support_feat
        else:
            support_x = support_x.view(self.n_way * self.k_shot, self.in_channels, self.n_points)
            support_feat = self.getFeatures(support_x)

        if cached_query_feat is not None:
            query_feat = cached_query_feat
        else:
            query_feat = self.getFeatures(query_x)

        support_feat = support_feat.view(self.n_way, self.k_shot, self.feat_dim, self.n_points)
        query_feat = query_feat.transpose(1, 2).contiguous().view(-1, self.feat_dim)

        fg_mask = support_y
        bg_mask = torch.logical_not(support_y)

        fg_prototypes, fg_labels = self.getForegroundPrototypes(support_feat, fg_mask, k=self.n_subprototypes)
        bg_prototype, bg_labels = self.getBackgroundPrototypes(support_feat, bg_mask, k=self.n_subprototypes)

        if bg_prototype is not None and bg_labels is not None:
            prototypes = torch.cat((bg_prototype, fg_prototypes), dim=0)
            prototype_labels = torch.cat((bg_labels, fg_labels), dim=0)
        else:
            prototypes = fg_prototypes
            prototype_labels = fg_labels
        self.num_prototypes = prototypes.shape[0]

        self.num_nodes = self.num_prototypes + query_feat.shape[0]
        Y = torch.zeros(self.num_nodes, self.n_classes, device=device)
        Y[:self.num_prototypes] = prototype_labels

        node_feat = torch.cat((prototypes, query_feat), dim=0)
        A = self.calculateLocalConstrainedAffinity(node_feat, k=self.k_connect)
        Z = self.label_propagate(A, Y)

        query_pred = Z[self.num_prototypes:, :]
        query_pred = query_pred.view(-1, query_y.shape[1], self.n_classes).transpose(1, 2)
        loss = self.computeCrossEntropyLoss(query_pred, query_y)
        return query_pred, loss

    def _forward_bayesian(self, support_x, support_y, query_x, query_y):
        device = support_x.device
        support_x_flat = support_x.view(self.n_way * self.k_shot, self.in_channels, self.n_points)
        support_feat = self.getFeatures(support_x_flat)
        support_feat = support_feat.transpose(1, 2).contiguous().view(-1, self.feat_dim)

        query_feat = self.getFeatures(query_x)
        query_feat_flat = query_feat.transpose(1, 2).contiguous().view(-1, self.feat_dim)

        support_labels_full = torch.zeros(
            support_feat.size(0), dtype=torch.long, device=device
        )
        for i in range(self.n_way):
            start_idx = i * self.k_shot * self.n_points
            mask_flat = support_y[i].view(-1)
            support_labels_full[start_idx:start_idx + mask_flat.size(0)][mask_flat] = i + 1

        self.bayesian_classifier.update_posterior(
            support_feat,
            support_labels_full,
            base_priors=getattr(self, 'base_priors', None),
            adapter=self.distribution_adapter if self.use_adapter else None
        )

        logits = self.bayesian_classifier.predict(query_feat_flat)
        query_pred = logits.view(-1, query_y.shape[1], self.n_classes).transpose(1, 2)
        loss = self.computeCrossEntropyLoss(query_pred, query_y)
        return query_pred, loss

    def getFeatures(self, x):
        if self.use_attention:
            feat_level1, feat_level2 = self.encoder(x)
            feat_level3 = self.base_learner(feat_level2)
            att_feat = self.att_learner(feat_level2)
            raw_feat = torch.cat((feat_level1, att_feat, feat_level3), dim=1)
        else:
            feat_level1, feat_level2 = self.encoder(x)
            feat_level3 = self.base_learner(feat_level2)
            map_feat = self.linear_mapper(feat_level2)
            raw_feat = torch.cat((feat_level1, map_feat, feat_level3), dim=1)

        B, D_raw, N = raw_feat.shape
        raw_feat = raw_feat.transpose(1, 2)  # [B, N, D_raw]
        proj_feat = self.feature_projector(raw_feat)  # [B, N, D_new]
        return proj_feat.transpose(1, 2).contiguous()  # [B, D_new, N]

    def getMutiplePrototypes(self, feat, k):
        n = feat.shape[0]
        if n == 0:
            return feat
        ratio = min(k / n, 1.0)
        if ratio < 1:
            fps_index = fps(feat, None, ratio=ratio, random_start=False).unique()
            if fps_index.numel() == 0:
                return feat[:1]
            farthest_seeds = feat[fps_index]
            distances = F.pairwise_distance(feat.unsqueeze(1), farthest_seeds.unsqueeze(0), p=2)
            assignments = torch.argmin(distances, dim=1)
            num_prototypes = farthest_seeds.size(0)
            prototypes = torch.zeros((num_prototypes, self.feat_dim), device=feat.device)
            counts = torch.zeros(num_prototypes, device=feat.device)
            for i in range(feat.size(0)):
                idx = assignments[i]
                prototypes[idx] += feat[i]
                counts[idx] += 1
            prototypes = prototypes / (counts.unsqueeze(1) + 1e-8)
            return prototypes
        else:
            return feat

    def getForegroundPrototypes(self, feats, masks, k=100):
        prototypes = []
        labels = []
        device = feats.device
        for i in range(self.n_way):
            feat = feats[i].permute(1, 2, 0).reshape(-1, self.feat_dim)
            mask_flat = masks[i].view(-1)
            if mask_flat.sum() == 0:
                continue
            feat_fg = feat[mask_flat]
            class_prototypes = self.getMutiplePrototypes(feat_fg, k)
            prototypes.append(class_prototypes)
            class_labels = torch.zeros(class_prototypes.shape[0], self.n_classes, device=device)
            class_labels[:, i + 1] = 1
            labels.append(class_labels)

        if prototypes:
            prototypes = torch.cat(prototypes, dim=0)
            labels = torch.cat(labels, dim=0)
        else:
            prototypes = torch.empty(0, self.feat_dim, device=device)
            labels = torch.empty(0, self.n_classes, device=device)
        return prototypes, labels

    def getBackgroundPrototypes(self, feats, masks, k=100):
        B, S, D, P = feats.shape
        feats_flat = feats.permute(0, 1, 3, 2).reshape(-1, D)
        masks_flat = masks.reshape(-1)
        if masks_flat.sum() == 0:
            return None, None
        feat_bg = feats_flat[masks_flat]
        prototypes = self.getMutiplePrototypes(feat_bg, k)
        labels = torch.zeros(prototypes.shape[0], self.n_classes, device=feats.device)
        labels[:, 0] = 1
        return prototypes, labels

    def calculateLocalConstrainedAffinity(self, node_feat, k=200, method='gaussian'):
        X = node_feat.detach().cpu().numpy()
        N = X.shape[0]
        if N <= 1:
            return torch.zeros(N, N, dtype=torch.float, device=node_feat.device)

        k_eff = min(k, N - 1)
        nbrs = NearestNeighbors(n_neighbors=k_eff + 1, algorithm='auto').fit(X)
        distances, indices = nbrs.kneighbors(X)
        I = torch.from_numpy(indices[:, 1:]).to(node_feat.device)

        knn_idx = I.unsqueeze(-1).expand(-1, -1, self.feat_dim).reshape(-1, self.feat_dim)
        knn_feat = torch.gather(node_feat, dim=0, index=knn_idx).reshape(N, k_eff, self.feat_dim)

        if method == 'gaussian':
            dist = F.pairwise_distance(node_feat.unsqueeze(1), knn_feat, p=2)
            knn_similarity = torch.exp(-0.5 * (dist / self.sigma) ** 2)
        else:
            raise NotImplementedError

        A = torch.zeros(N, N, dtype=torch.float, device=node_feat.device)
        A.scatter_(1, I, knn_similarity)
        A = A + A.t()
        A.fill_diagonal_(0.0)
        return A

    def label_propagate(self, A, Y, alpha=0.99):
        eps = 1e-8
        D = A.sum(1)
        D_sqrt_inv = torch.sqrt(1.0 / (D + eps))
        D_sqrt_inv = torch.diag_embed(D_sqrt_inv).to(A.device)
        S = D_sqrt_inv @ A @ D_sqrt_inv
        M = torch.eye(self.num_nodes, device=A.device) - alpha * S + eps * torch.eye(self.num_nodes, device=A.device)
        Z = torch.linalg.solve(M, Y)
        return Z

    def computeCrossEntropyLoss(self, query_logits, query_labels):
        return F.cross_entropy(query_logits, query_labels)
```

```python
%%writefile /content/attMPTI/models/mpti.py

""" Multi-prototype transductive inference with Bayesian adaptation for distribution shift

Author: Zhao Na, 2020 — Extended with Bayesian + Distribution Shift Adapter (Optimized for Colab + Trainable)
"""
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch_cluster import fps

from models.dgcnn import DGCNN
from models.attention import SelfAttention
from models.bayesian_classifier import BayesianClassifier
from models.adapter import DistributionShiftAdapter
from sklearn.neighbors import NearestNeighbors


class BaseLearner(nn.Module):
    """The class for inner loop."""
    def __init__(self, in_channels, params):
        super(BaseLearner, self).__init__()
        self.num_convs = len(params)
        self.convs = nn.ModuleList()
        for i in range(self.num_convs):
            in_dim = in_channels if i == 0 else params[i-1]
            self.convs.append(nn.Sequential(
                nn.Conv1d(in_dim, params[i], 1),
                nn.BatchNorm1d(params[i])
            ))

    def forward(self, x):
        for i in range(self.num_convs):
            x = self.convs[i](x)
            if i != self.num_convs - 1:
                x = F.relu(x)
        return x


class MultiPrototypeTransductiveInference(nn.Module):
    def __init__(self, args):
        super(MultiPrototypeTransductiveInference, self).__init__()
        self.n_way = args.n_way
        self.k_shot = args.k_shot
        self.in_channels = args.pc_in_dim
        self.n_points = args.pc_npts
        self.use_attention = args.use_attention
        self.n_subprototypes = args.n_subprototypes
        self.k_connect = args.k_connect
        self.sigma = args.sigma
        self.n_classes = self.n_way + 1  # +1 for background

        # Backbone
        self.encoder = DGCNN(args.edgeconv_widths, args.dgcnn_mlp_widths, args.pc_in_dim, k=args.dgcnn_k)
        self.base_learner = BaseLearner(args.dgcnn_mlp_widths[-1], args.base_widths)

        if self.use_attention:
            self.att_learner = SelfAttention(args.dgcnn_mlp_widths[-1], args.output_dim)
        else:
            self.linear_mapper = nn.Conv1d(args.dgcnn_mlp_widths[-1], args.output_dim, 1, bias=False)

        raw_feat_dim = args.edgeconv_widths[0][-1] + args.output_dim + args.base_widths[-1]
        self.projected_dim = getattr(args, 'projected_dim', 128)
        self.feature_projector = nn.Linear(raw_feat_dim, self.projected_dim)
        self.feat_dim = self.projected_dim

        # Bayesian components
        self.use_bayesian = getattr(args, 'use_bayesian', False)
        if self.use_bayesian:
            self.bayesian_classifier = BayesianClassifier(
                num_classes=self.n_classes,
                embedding_dim=self.feat_dim,
                beta=getattr(args, 'beta', 8.0),
                precision=getattr(args, 'precision', 1.0)
            )
            self.use_adapter = getattr(args, 'use_adapter', False)
            if self.use_adapter:
                self.distribution_adapter = DistributionShiftAdapter(self.feat_dim)

    def forward(self, support_x, support_y, query_x, query_y):
        if self.use_bayesian:
            return self._forward_bayesian(support_x, support_y, query_x, query_y)
        else:
            return self._forward_label_propagation(support_x, support_y, query_x, query_y)

    def _forward_label_propagation(self, support_x, support_y, query_x, query_y,
                                   cached_support_feat=None, cached_query_feat=None):
        device = support_x.device

        if cached_support_feat is not None:
            support_feat = cached_support_feat
        else:
            support_x = support_x.view(self.n_way * self.k_shot, self.in_channels, self.n_points)
            support_feat = self.getFeatures(support_x)

        if cached_query_feat is not None:
            query_feat = cached_query_feat
        else:
            query_feat = self.getFeatures(query_x)

        support_feat = support_feat.view(self.n_way, self.k_shot, self.feat_dim, self.n_points)
        query_feat = query_feat.transpose(1, 2).contiguous().view(-1, self.feat_dim)

        fg_mask = support_y
        bg_mask = torch.logical_not(support_y)

        fg_prototypes, fg_labels = self.getForegroundPrototypes(support_feat, fg_mask, k=self.n_subprototypes)
        bg_prototype, bg_labels = self.getBackgroundPrototypes(support_feat, bg_mask, k=self.n_subprototypes)

        if bg_prototype is not None and bg_labels is not None:
            prototypes = torch.cat((bg_prototype, fg_prototypes), dim=0)
            prototype_labels = torch.cat((bg_labels, fg_labels), dim=0)
        else:
            prototypes = fg_prototypes
            prototype_labels = fg_labels
        self.num_prototypes = prototypes.shape[0]

        self.num_nodes = self.num_prototypes + query_feat.shape[0]
        Y = torch.zeros(self.num_nodes, self.n_classes, device=device)
        Y[:self.num_prototypes] = prototype_labels

        node_feat = torch.cat((prototypes, query_feat), dim=0)
        A = self.calculateLocalConstrainedAffinity(node_feat, k=self.k_connect)
        Z = self.label_propagate(A, Y)

        query_pred = Z[self.num_prototypes:, :]
        query_pred = query_pred.view(-1, query_y.shape[1], self.n_classes).transpose(1, 2)
        loss = self.computeCrossEntropyLoss(query_pred, query_y)
        return query_pred, loss

    def _forward_bayesian(self, support_x, support_y, query_x, query_y):
        device = support_x.device
        support_x_flat = support_x.view(self.n_way * self.k_shot, self.in_channels, self.n_points)
        support_feat = self.getFeatures(support_x_flat)
        support_feat = support_feat.transpose(1, 2).contiguous().view(-1, self.feat_dim)

        query_feat = self.getFeatures(query_x)
        query_feat_flat = query_feat.transpose(1, 2).contiguous().view(-1, self.feat_dim)

        # Reset classifier buffers
        self.bayesian_classifier.mean.zero_()
        self.bayesian_classifier.var.fill_(1.0)

        # Process each novel class separately
        for c in range(self.n_way):
            class_label = c + 1  # foreground classes: 1 to n_way
            start_idx = c * self.k_shot * self.n_points
            end_idx = (c + 1) * self.k_shot * self.n_points
            mask_local = support_y[c].view(-1)  # [k_shot * N]
            global_mask = torch.zeros(support_feat.size(0), dtype=torch.bool, device=device)
            global_mask[start_idx:start_idx + mask_local.numel()] = mask_local

            if global_mask.sum() == 0:
                continue

            class_support_feat = support_feat[global_mask]  # [M, D]
            Nc = class_support_feat.size(0)

            # Construct hybrid prior (base + self)
            mu_hybrid, var_hybrid = self.bayesian_classifier.construct_prior_for_novel_class(
                class_support_feat,
                base_priors=getattr(self, 'base_priors', None),
                beta=self.bayesian_classifier.beta,
                mode='hybrid',
                lambda_mix=0.4
            )

            # Apply adapter if enabled
            if self.use_adapter:
                with torch.no_grad():
                    Sigma_hybrid = torch.diag(var_hybrid)
                    mu_prior, Sigma_prior = self.distribution_adapter(mu_hybrid, Sigma_hybrid)
                    var_prior = torch.diagonal(Sigma_prior)
            else:
                mu_prior = mu_hybrid
                var_prior = var_hybrid

            # Update posterior: use Nc as likelihood precision
            support_mean = class_support_feat.mean(dim=0)
            prec_prior = 1.0 / var_prior
            prec_likelihood = Nc  # ⚠️ key fix: precision proportional to sample count

            posterior_prec = prec_prior + prec_likelihood
            posterior_mean = (prec_prior * mu_prior + prec_likelihood * support_mean) / posterior_prec
            posterior_var = 1.0 / posterior_prec

            # Numerical stability
            posterior_mean = torch.nan_to_num(posterior_mean, nan=0.0, posinf=1e6, neginf=-1e6)
            posterior_var = torch.clamp(posterior_var, min=1e-8, max=1e6)

            # 🔥 CRITICAL: detach before assigning to buffer
            self.bayesian_classifier.mean[class_label] = posterior_mean.detach()
            self.bayesian_classifier.var[class_label] = posterior_var.detach()

        # 🔥 Initialize background class (label 0) with neutral prior — also detached
        self.bayesian_classifier.mean[0] = torch.zeros_like(self.bayesian_classifier.mean[0]).detach()
        self.bayesian_classifier.var[0] = torch.ones_like(self.bayesian_classifier.var[0]).detach()

        # Predict on query
        logits = self.bayesian_classifier.predict(query_feat_flat)
        query_pred = logits.view(-1, query_y.shape[1], self.n_classes).transpose(1, 2)
        loss = self.computeCrossEntropyLoss(query_pred, query_y)
        return query_pred, loss

    def getFeatures(self, x):
        if self.use_attention:
            feat_level1, feat_level2 = self.encoder(x)
            feat_level3 = self.base_learner(feat_level2)
            att_feat = self.att_learner(feat_level2)
            raw_feat = torch.cat((feat_level1, att_feat, feat_level3), dim=1)
        else:
            feat_level1, feat_level2 = self.encoder(x)
            feat_level3 = self.base_learner(feat_level2)
            map_feat = self.linear_mapper(feat_level2)
            raw_feat = torch.cat((feat_level1, map_feat, feat_level3), dim=1)

        B, D_raw, N = raw_feat.shape
        raw_feat = raw_feat.transpose(1, 2)  # [B, N, D_raw]
        proj_feat = self.feature_projector(raw_feat)  # [B, N, D_new]
        return proj_feat.transpose(1, 2).contiguous()  # [B, D_new, N]

    def getMutiplePrototypes(self, feat, k):
        n = feat.shape[0]
        if n == 0:
            return feat
        ratio = min(k / n, 1.0)
        if ratio < 1:
            fps_index = fps(feat, None, ratio=ratio, random_start=False).unique()
            if fps_index.numel() == 0:
                return feat[:1]
            farthest_seeds = feat[fps_index]
            distances = F.pairwise_distance(feat.unsqueeze(1), farthest_seeds.unsqueeze(0), p=2)
            assignments = torch.argmin(distances, dim=1)
            num_prototypes = farthest_seeds.size(0)
            prototypes = torch.zeros((num_prototypes, self.feat_dim), device=feat.device)
            counts = torch.zeros(num_prototypes, device=feat.device)
            for i in range(feat.size(0)):
                idx = assignments[i]
                prototypes[idx] += feat[i]
                counts[idx] += 1
            prototypes = prototypes / (counts.unsqueeze(1) + 1e-8)
            return prototypes
        else:
            return feat

    def getForegroundPrototypes(self, feats, masks, k=100):
        prototypes = []
        labels = []
        device = feats.device
        for i in range(self.n_way):
            feat = feats[i].permute(1, 2, 0).reshape(-1, self.feat_dim)
            mask_flat = masks[i].view(-1)
            if mask_flat.sum() == 0:
                continue
            feat_fg = feat[mask_flat]
            class_prototypes = self.getMutiplePrototypes(feat_fg, k)
            prototypes.append(class_prototypes)
            class_labels = torch.zeros(class_prototypes.shape[0], self.n_classes, device=device)
            class_labels[:, i + 1] = 1
            labels.append(class_labels)

        if prototypes:
            prototypes = torch.cat(prototypes, dim=0)
            labels = torch.cat(labels, dim=0)
        else:
            prototypes = torch.empty(0, self.feat_dim, device=device)
            labels = torch.empty(0, self.n_classes, device=device)
        return prototypes, labels

    def getBackgroundPrototypes(self, feats, masks, k=100):
        B, S, D, P = feats.shape
        feats_flat = feats.permute(0, 1, 3, 2).reshape(-1, D)
        masks_flat = masks.reshape(-1)
        if masks_flat.sum() == 0:
            return None, None
        feat_bg = feats_flat[masks_flat]
        prototypes = self.getMutiplePrototypes(feat_bg, k)
        labels = torch.zeros(prototypes.shape[0], self.n_classes, device=feats.device)
        labels[:, 0] = 1
        return prototypes, labels

    def calculateLocalConstrainedAffinity(self, node_feat, k=200, method='gaussian'):
        X = node_feat.detach().cpu().numpy()
        N = X.shape[0]
        if N <= 1:
            return torch.zeros(N, N, dtype=torch.float, device=node_feat.device)

        k_eff = min(k, N - 1)
        nbrs = NearestNeighbors(n_neighbors=k_eff + 1, algorithm='auto').fit(X)
        distances, indices = nbrs.kneighbors(X)
        I = torch.from_numpy(indices[:, 1:]).to(node_feat.device)

        knn_idx = I.unsqueeze(-1).expand(-1, -1, self.feat_dim).reshape(-1, self.feat_dim)
        knn_feat = torch.gather(node_feat, dim=0, index=knn_idx).reshape(N, k_eff, self.feat_dim)

        if method == 'gaussian':
            dist = F.pairwise_distance(node_feat.unsqueeze(1), knn_feat, p=2)
            knn_similarity = torch.exp(-0.5 * (dist / self.sigma) ** 2)
        else:
            raise NotImplementedError

        A = torch.zeros(N, N, dtype=torch.float, device=node_feat.device)
        A.scatter_(1, I, knn_similarity)
        A = A + A.t()
        A.fill_diagonal_(0.0)
        return A

    def label_propagate(self, A, Y, alpha=0.99):
        eps = 1e-8
        D = A.sum(1)
        D_sqrt_inv = torch.sqrt(1.0 / (D + eps))
        D_sqrt_inv = torch.diag_embed(D_sqrt_inv).to(A.device)
        S = D_sqrt_inv @ A @ D_sqrt_inv
        M = torch.eye(self.num_nodes, device=A.device) - alpha * S + eps * torch.eye(self.num_nodes, device=A.device)
        Z = torch.linalg.solve(M, Y)
        return Z

    def computeCrossEntropyLoss(self, query_logits, query_labels):
        return F.cross_entropy(query_logits, query_labels)
```

```python
%%writefile /content/attMPTI/models/mpti.py

""" Multi-prototype transductive inference with Bayesian adaptation for distribution shift

Author: Zhao Na, 2020 — Extended with Bayesian + Distribution Shift Adapter (Optimized for Colab + Trainable)
"""
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch_cluster import fps

from models.dgcnn import DGCNN
from models.attention import SelfAttention
from models.bayesian_classifier import BayesianClassifier
from models.adapter import DistributionShiftAdapter
from sklearn.neighbors import NearestNeighbors


class BaseLearner(nn.Module):
    """The class for inner loop."""
    def __init__(self, in_channels, params):
        super(BaseLearner, self).__init__()
        self.num_convs = len(params)
        self.convs = nn.ModuleList()
        for i in range(self.num_convs):
            in_dim = in_channels if i == 0 else params[i-1]
            self.convs.append(nn.Sequential(
                nn.Conv1d(in_dim, params[i], 1),
                nn.BatchNorm1d(params[i])
            ))

    def forward(self, x):
        for i in range(self.num_convs):
            x = self.convs[i](x)
            if i != self.num_convs - 1:
                x = F.relu(x)
        return x


class MultiPrototypeTransductiveInference(nn.Module):
    def __init__(self, args):
        super(MultiPrototypeTransductiveInference, self).__init__()
        self.n_way = args.n_way
        self.k_shot = args.k_shot
        self.in_channels = args.pc_in_dim
        self.n_points = args.pc_npts
        self.use_attention = args.use_attention
        self.n_subprototypes = args.n_subprototypes
        self.k_connect = args.k_connect
        self.sigma = args.sigma
        # 🔥 حذف کلاس background: فقط n_way کلاس نو
        self.n_classes = self.n_way

        # Backbone
        self.encoder = DGCNN(args.edgeconv_widths, args.dgcnn_mlp_widths, args.pc_in_dim, k=args.dgcnn_k)
        self.base_learner = BaseLearner(args.dgcnn_mlp_widths[-1], args.base_widths)

        if self.use_attention:
            self.att_learner = SelfAttention(args.dgcnn_mlp_widths[-1], args.output_dim)
        else:
            self.linear_mapper = nn.Conv1d(args.dgcnn_mlp_widths[-1], args.output_dim, 1, bias=False)

        raw_feat_dim = args.edgeconv_widths[0][-1] + args.output_dim + args.base_widths[-1]
        self.projected_dim = getattr(args, 'projected_dim', 128)
        self.feature_projector = nn.Linear(raw_feat_dim, self.projected_dim)
        self.feat_dim = self.projected_dim

        # Bayesian components
        self.use_bayesian = getattr(args, 'use_bayesian', False)
        if self.use_bayesian:
            self.bayesian_classifier = BayesianClassifier(
                num_classes=self.n_classes,  # ✅ n_way, not n_way+1
                embedding_dim=self.feat_dim,
                beta=getattr(args, 'beta', 8.0),
                temperature=getattr(args, 'temperature', 1.0)
            )
            self.use_adapter = getattr(args, 'use_adapter', False)
            if self.use_adapter:
                self.distribution_adapter = DistributionShiftAdapter(self.feat_dim)

    def forward(self, support_x, support_y, query_x, query_y):
        if self.use_bayesian:
            return self._forward_bayesian(support_x, support_y, query_x, query_y)
        else:
            return self._forward_label_propagation(support_x, support_y, query_x, query_y)

    def _forward_label_propagation(self, support_x, support_y, query_x, query_y,
                                   cached_support_feat=None, cached_query_feat=None):
        device = support_x.device

        if cached_support_feat is not None:
            support_feat = cached_support_feat
        else:
            support_x = support_x.view(self.n_way * self.k_shot, self.in_channels, self.n_points)
            support_feat = self.getFeatures(support_x)

        if cached_query_feat is not None:
            query_feat = cached_query_feat
        else:
            query_feat = self.getFeatures(query_x)

        support_feat = support_feat.view(self.n_way, self.k_shot, self.feat_dim, self.n_points)
        query_feat = query_feat.transpose(1, 2).contiguous().view(-1, self.feat_dim)

        # 🔥 بدون background: تمام maskها foreground هستند
        fg_prototypes, fg_labels = self.getForegroundPrototypes(support_feat, support_y, k=self.n_subprototypes)
        prototypes = fg_prototypes
        prototype_labels = fg_labels
        self.num_prototypes = prototypes.shape[0]

        self.num_nodes = self.num_prototypes + query_feat.shape[0]
        Y = torch.zeros(self.num_nodes, self.n_classes, device=device)
        Y[:self.num_prototypes] = prototype_labels

        node_feat = torch.cat((prototypes, query_feat), dim=0)
        A = self.calculateLocalConstrainedAffinity(node_feat, k=self.k_connect)
        Z = self.label_propagate(A, Y)

        query_pred = Z[self.num_prototypes:, :]
        query_pred = query_pred.view(-1, query_y.shape[1], self.n_classes).transpose(1, 2)
        loss = self.computeCrossEntropyLoss(query_pred, query_y)
        return query_pred, loss

    def _forward_bayesian(self, support_x, support_y, query_x, query_y):
        device = support_x.device

        # Handle support data
        if support_x.dim() == 5:  # [B, n_way, k_shot, C, N]
            support_x = support_x.squeeze(0)  # [n_way, k_shot, C, N]
            support_y = support_y.squeeze(0)  # [n_way, k_shot, N]
        elif support_x.dim() == 4:  # [n_way, k_shot, N, C]
            support_x = support_x.permute(0, 1, 3, 2)  # [n_way, k_shot, C, N]
        else:
            raise ValueError(f"Unexpected support_x shape: {support_x.shape}")

        # Process support features
        support_x_flat = support_x.view(self.n_way * self.k_shot, self.in_channels, self.n_points)
        support_feat = self.getFeatures(support_x_flat)  # [B, D, N]
        support_feat = support_feat.transpose(1, 2).contiguous().view(-1, self.feat_dim)  # [M, D]
        support_feat = support_feat / (support_feat.norm(dim=1, keepdim=True) + 1e-8)

        # Process query features
        if query_x.dim() == 2:  # [C, N]
            query_x = query_x.unsqueeze(0)  # [1, C, N]
        elif query_x.dim() == 4:  # [B, Q, N, C]
            B, Q, N, C = query_x.shape
            query_x = query_x.view(B, Q * N, C).transpose(1, 2)  # [B, C, Q*N]
        if query_x.dim() != 3 or query_x.shape[1] != self.in_channels:
            query_x = query_x.view(-1, self.in_channels).transpose(0, 1).unsqueeze(0)

        query_feat = self.getFeatures(query_x)  # [1, D, N_total]
        query_feat_flat = query_feat.transpose(1, 2).contiguous().view(-1, self.feat_dim)  # [Q, D]
        query_feat_flat = query_feat_flat / (query_feat_flat.norm(dim=1, keepdim=True) + 1e-8)

        # 🔥 فقط پروتوتایپ‌های foreground (بدون background)
        prototypes = []

        for c in range(self.n_way):
            mask_local = support_y[c].view(-1)  # [k_shot * N]
            start_idx = c * self.k_shot * self.n_points
            global_mask = torch.zeros(support_feat.size(0), dtype=torch.bool, device=device)
            global_mask[start_idx:start_idx + mask_local.numel()] = mask_local

            if global_mask.sum() == 0:
                mu_self = torch.zeros(self.feat_dim, device=device)
            else:
                class_support_feat = support_feat[global_mask]
                mu_self = class_support_feat.mean(dim=0)

            mu_self = mu_self / (mu_self.norm() + 1e-8)

            # Construct prior (differentiable!)
            if hasattr(self, 'base_priors') and len(self.base_priors) > 0:
                mu_prior = self.bayesian_classifier.construct_prior_for_novel_class(
                    class_support_feat if global_mask.sum() > 0 else mu_self.unsqueeze(0),
                    base_priors=self.base_priors,
                    beta=self.bayesian_classifier.beta,
                    lambda_mix=0.4
                )
            else:
                mu_prior = mu_self

            # Apply adapter (differentiable!)
            if self.use_adapter:
                mu_prior = self.distribution_adapter(mu_prior)

            prototypes.append(mu_prior.unsqueeze(0))  # [1, D]

        # Stack only novel class prototypes: [n_way, D]
        class_means = torch.cat(prototypes, dim=0)  # [n_way, D]
        class_means = class_means / (class_means.norm(dim=1, keepdim=True) + 1e-8)

        # Predict: cosine similarity
        query_norm = query_feat_flat / (query_feat_flat.norm(dim=1, keepdim=True) + 1e-8)
        cos_sim = torch.mm(query_norm, class_means.t())  # [Q, n_way]
        logits = cos_sim / self.bayesian_classifier.temperature  # [Q, n_way]

        # Flatten query_y to [Q]
        if query_y.dim() >= 2:
            query_y = query_y.view(-1)  # [Q]
        elif query_y.dim() == 1:
            pass

        # 🔥 برچسب‌های محلی از 0 تا n_way-1 هستند
        assert query_y.max() < self.n_classes, f"Label {query_y.max()} >= num_classes {self.n_classes}"
        assert query_y.min() >= 0, f"Negative label found: {query_y.min()}"

        loss = F.cross_entropy(logits, query_y)

        # For compatibility with evaluator (expects [1, C, Q])
        query_pred = logits.transpose(0, 1).unsqueeze(0)  # [1, n_way, Q]

        return query_pred, loss

    def getFeatures(self, x):
        if self.use_attention:
            feat_level1, feat_level2 = self.encoder(x)
            feat_level3 = self.base_learner(feat_level2)
            att_feat = self.att_learner(feat_level2)
            raw_feat = torch.cat((feat_level1, att_feat, feat_level3), dim=1)
        else:
            feat_level1, feat_level2 = self.encoder(x)
            feat_level3 = self.base_learner(feat_level2)
            map_feat = self.linear_mapper(feat_level2)
            raw_feat = torch.cat((feat_level1, map_feat, feat_level3), dim=1)

        B, D_raw, N = raw_feat.shape
        raw_feat = raw_feat.transpose(1, 2)  # [B, N, D_raw]
        proj_feat = self.feature_projector(raw_feat)  # [B, N, D_new]
        return proj_feat.transpose(1, 2).contiguous()  # [B, D_new, N]

    def getMutiplePrototypes(self, feat, k):
        n = feat.shape[0]
        if n == 0:
            return feat
        ratio = min(k / n, 1.0)
        if ratio < 1:
            fps_index = fps(feat, None, ratio=ratio, random_start=False).unique()
            if fps_index.numel() == 0:
                return feat[:1]
            farthest_seeds = feat[fps_index]
            distances = F.pairwise_distance(feat.unsqueeze(1), farthest_seeds.unsqueeze(0), p=2)
            assignments = torch.argmin(distances, dim=1)
            num_prototypes = farthest_seeds.size(0)
            prototypes = torch.zeros((num_prototypes, self.feat_dim), device=feat.device)
            counts = torch.zeros(num_prototypes, device=feat.device)
            for i in range(feat.size(0)):
                idx = assignments[i]
                prototypes[idx] += feat[i]
                counts[idx] += 1
            prototypes = prototypes / (counts.unsqueeze(1) + 1e-8)
            return prototypes
        else:
            return feat

    def getForegroundPrototypes(self, feats, masks, k=100):
        prototypes = []
        labels = []
        device = feats.device
        for i in range(self.n_way):
            feat = feats[i].permute(1, 2, 0).reshape(-1, self.feat_dim)
            mask_flat = masks[i].view(-1)
            if mask_flat.sum() == 0:
                continue
            feat_fg = feat[mask_flat]
            class_prototypes = self.getMutiplePrototypes(feat_fg, k)
            prototypes.append(class_prototypes)
            # 🔥 بدون offset: کلاس محلی i به جای i+1
            class_labels = torch.zeros(class_prototypes.shape[0], self.n_classes, device=device)
            class_labels[:, i] = 1  # ✅ تغییر کلیدی: i به جای i+1
            labels.append(class_labels)

        if prototypes:
            prototypes = torch.cat(prototypes, dim=0)
            labels = torch.cat(labels, dim=0)
        else:
            prototypes = torch.empty(0, self.feat_dim, device=device)
            labels = torch.empty(0, self.n_classes, device=device)
        return prototypes, labels

    # 🔥 حذف getBackgroundPrototypes — دیگر استفاده نمی‌شود

    def calculateLocalConstrainedAffinity(self, node_feat, k=200, method='gaussian'):
        X = node_feat.detach().cpu().numpy()
        N = X.shape[0]
        if N <= 1:
            return torch.zeros(N, N, dtype=torch.float, device=node_feat.device)

        k_eff = min(k, N - 1)
        nbrs = NearestNeighbors(n_neighbors=k_eff + 1, algorithm='auto').fit(X)
        distances, indices = nbrs.kneighbors(X)
        I = torch.from_numpy(indices[:, 1:]).to(node_feat.device)

        knn_idx = I.unsqueeze(-1).expand(-1, -1, self.feat_dim).reshape(-1, self.feat_dim)
        knn_feat = torch.gather(node_feat, dim=0, index=knn_idx).reshape(N, k_eff, self.feat_dim)

        if method == 'gaussian':
            dist = F.pairwise_distance(node_feat.unsqueeze(1), knn_feat, p=2)
            knn_similarity = torch.exp(-0.5 * (dist / self.sigma) ** 2)
        else:
            raise NotImplementedError

        A = torch.zeros(N, N, dtype=torch.float, device=node_feat.device)
        A.scatter_(1, I, knn_similarity)
        A = A + A.t()
        A.fill_diagonal_(0.0)
        return A

    def label_propagate(self, A, Y, alpha=0.99):
        eps = 1e-8
        D = A.sum(1)
        D_sqrt_inv = torch.sqrt(1.0 / (D + eps))
        D_sqrt_inv = torch.diag_embed(D_sqrt_inv).to(A.device)
        S = D_sqrt_inv @ A @ D_sqrt_inv
        M = torch.eye(self.num_nodes, device=A.device) - alpha * S + eps * torch.eye(self.num_nodes, device=A.device)
        Z = torch.linalg.solve(M, Y)
        return Z

    def computeCrossEntropyLoss(self, query_logits, query_labels):
        return F.cross_entropy(query_logits, query_labels)
```

```python
%%writefile /content/attMPTI/runs/eval.py

"""Evaluating functions for Few-shot 3D Point Cloud Semantic Segmentation

Author: Zhao Na, 2020 — Updated for Mean-Only Bayesian + Collate Fix
"""
import os
import numpy as np
from datetime import datetime

import torch
from torch.utils.data import DataLoader

from dataloaders.loader import MyTestDataset, batch_test_task_collate
from models.proto_learner import ProtoLearner
from models.mpti_learner import MPTILearner
from utils.cuda_util import cast_cuda
from utils.logger import init_logger


def evaluate_metric(logger, pred_labels_list, gt_labels_list, label2class_list, test_classes):
    """
    Compute IoU per episode, then average over episodes.
    Each episode has its own n_way classes.
    """
    assert len(pred_labels_list) == len(gt_labels_list) == len(label2class_list)

    logger.cprint('*****Test Classes (global): {0}*****'.format(test_classes))

    episode_ious = []

    for i in range(len(gt_labels_list)):
        batch_gt_labels = np.atleast_1d(gt_labels_list[i]).flatten()
        batch_pred_labels = np.atleast_1d(pred_labels_list[i]).flatten()
        label2class = label2class_list[i]

        # 🔥 اصلاح: هم‌تراز کردن طول‌ها در صورت نیاز
        if len(batch_gt_labels) != len(batch_pred_labels):
            min_len = min(len(batch_gt_labels), len(batch_pred_labels))
            batch_gt_labels = batch_gt_labels[:min_len]
            batch_pred_labels = batch_pred_labels[:min_len]

        n_way = len(label2class)
        NUM_CLASS = n_way + 1  # background (0) + n_way novel classes

        gt_classes = [0] * NUM_CLASS
        positive_classes = [0] * NUM_CLASS
        true_positive_classes = [0] * NUM_CLASS

        for idx in range(len(batch_gt_labels)):
            gt = int(batch_gt_labels[idx])
            pred = int(batch_pred_labels[idx])

            # Labels are local: 0=bg, 1..n_way = novel classes
            if gt < 0 or gt >= NUM_CLASS:
                continue  # skip invalid
            if pred < 0 or pred >= NUM_CLASS:
                pred = 0  # map invalid pred to background

            gt_classes[gt] += 1
            positive_classes[pred] += 1
            if gt == pred:
                true_positive_classes[gt] += 1

        iou_list = []
        for c in range(NUM_CLASS):
            denom = gt_classes[c] + positive_classes[c] - true_positive_classes[c]
            iou = true_positive_classes[c] / (denom + 1e-8)
            iou_list.append(iou)

        # Mean IoU over novel classes only (exclude background)
        if n_way > 0:
            mean_iou_episode = np.mean(iou_list[1:])
        else:
            mean_iou_episode = 0.0
        episode_ious.append(mean_iou_episode)

    mean_IoU = np.mean(episode_ious) if episode_ious else 0.0
    return mean_IoU


def test_few_shot(test_loader, learner, logger, test_classes):
    if len(test_loader) == 0:
        logger.cprint("⚠️  No validation episodes available!")
        return 0.0, 0.0

    total_loss = 0
    predicted_label_total = []
    gt_label_total = []
    label2class_total = []

    for batch_idx, episode_data in enumerate(test_loader):
        support_x, support_y, query_x, query_y, sampled_classes = episode_data

        data = (support_x, support_y, query_x, query_y)
        query_label = query_y

        if torch.cuda.is_available():
            data = tuple(x.cuda(non_blocking=True) for x in data)

        query_pred, loss, accuracy = learner.test(data)
        total_loss += loss.detach().item()

        # 🔥 پردازش پیش‌بینی: [1, C, Q] → [Q]
        query_pred_np = query_pred.cpu().detach().numpy()
        pred_labels = np.argmax(query_pred_np, axis=1).squeeze()

        # 🔥 پردازش برچسب واقعی
        query_label_np = query_label.cpu().numpy().squeeze()

        # اطمینان از 1D بودن
        pred_labels = np.atleast_1d(pred_labels)
        query_label_np = np.atleast_1d(query_label_np)

        # 🔥 هم‌تراز کردن طول‌ها (کلید اصلی رفع خطا)
        if len(pred_labels) != len(query_label_np):
            min_len = min(len(pred_labels), len(query_label_np))
            pred_labels = pred_labels[:min_len]
            query_label_np = query_label_np[:min_len]

        predicted_label_total.append(pred_labels)
        gt_label_total.append(query_label_np)
        label2class_total.append(sampled_classes)

        if torch.cuda.is_available():
            torch.cuda.empty_cache()

    mean_loss = total_loss / len(test_loader)
    mean_IoU = evaluate_metric(logger, predicted_label_total, gt_label_total, label2class_total, test_classes)
    return mean_loss, mean_IoU


def eval(args):
    logger = init_logger(args.log_dir, args)

    ckpt_path = args.model_checkpoint_path
    assert ckpt_path is not None, "model_checkpoint_path must be provided for evaluation"

    ckpt = torch.load(ckpt_path, map_location='cpu')

    if args.phase == 'protoeval':
        learner = ProtoLearner(args, mode='test')
    elif args.phase == 'mptieval':
        learner = MPTILearner(args, mode='test')
        if 'base_priors' in ckpt:
            device = next(learner.model.parameters()).device
            loaded_priors = {}
            sample_item = next(iter(ckpt['base_priors'].values()))
            if isinstance(sample_item, (tuple, list)) and len(sample_item) == 2:
                for k, (mu, var) in ckpt['base_priors'].items():
                    loaded_priors[k] = mu.to(device)
            else:
                for k, mu in ckpt['base_priors'].items():
                    loaded_priors[k] = mu.to(device)
            learner.model.base_priors = loaded_priors
            logger.cprint("✅ Loaded base_priors for Bayesian inference.")
        else:
            logger.cprint("⚠️ Warning: base_priors not found. Using 'self' prior.")

    TEST_DATASET = MyTestDataset(args.data_path, args.dataset, cvfold=args.cvfold,
                                 num_episode_per_comb=args.n_episode_test,
                                 n_way=args.n_way, k_shot=args.k_shot, n_queries=args.n_queries,
                                 num_point=args.pc_npts, pc_attribs=args.pc_attribs, mode='test')
    TEST_CLASSES = list(TEST_DATASET.classes)

    TEST_LOADER = DataLoader(
        TEST_DATASET,
        batch_size=1,
        shuffle=False,
        collate_fn=None,
        num_workers=0
    )

    test_loss, mean_IoU = test_few_shot(TEST_LOADER, learner, logger, TEST_CLASSES)
    logger.cprint('\n=====[TEST] Loss: %.4f | Mean IoU: %.4f =====\n' % (test_loss, mean_IoU))
```

```python
%%writefile /content/attMPTI/dataloaders/loader.py
""" Data Loader for Generating Tasks

Author: Zhao Na, 2020 — Optimized for Colab (lightweight + Bayesian-compatible)
"""
import os
import random
import math
import glob
import numpy as np
import h5py as h5
import transforms3d
from itertools import combinations

import torch
from torch.utils.data import Dataset


def sample_K_pointclouds(data_path, num_point, pc_attribs, pc_augm, pc_augm_config,
                         scan_names, sampled_class, sampled_classes, is_support=False):
    '''sample K pointclouds and the corresponding labels for one class (one_way)'''
    ptclouds = []
    labels = []
    for scan_name in scan_names:
        ptcloud, label = sample_pointcloud(data_path, num_point, pc_attribs, pc_augm, pc_augm_config,
                                           scan_name, sampled_classes, sampled_class, support=is_support)
        ptclouds.append(ptcloud)
        labels.append(label)

    ptclouds = np.stack(ptclouds, axis=0)
    labels = np.stack(labels, axis=0)

    return ptclouds, labels


def sample_pointcloud(data_path, num_point, pc_attribs, pc_augm, pc_augm_config, scan_name,
                      sampled_classes, sampled_class=0, support=False, random_sample=False):
    sampled_classes = list(sampled_classes)
    data = np.load(os.path.join(data_path, 'data', '%s.npy' % scan_name))
    N = data.shape[0]  # number of points in this scan

    if random_sample:
        sampled_point_inds = np.random.choice(np.arange(N), num_point, replace=(N < num_point))
    else:
        # If this point cloud is for support/query set, make sure that the sampled points contain target class
        valid_point_inds = np.nonzero(data[:, 6] == sampled_class)[0]  # indices of points belonging to the sampled class

        if len(valid_point_inds) == 0:
            # Fallback: random sample if no valid points
            sampled_point_inds = np.random.choice(np.arange(N), num_point, replace=(N < num_point))
        else:
            if N < num_point:
                sampled_valid_point_num = len(valid_point_inds)
            else:
                valid_ratio = len(valid_point_inds) / float(N)
                sampled_valid_point_num = max(1, int(valid_ratio * num_point))  # ⚠️ حداقل 1 نقطه

            sampled_valid_point_inds = np.random.choice(valid_point_inds, sampled_valid_point_num, replace=False)
            sampled_other_point_inds = np.random.choice(np.arange(N), num_point - sampled_valid_point_num,
                                                        replace=(N < num_point))
            sampled_point_inds = np.concatenate([sampled_valid_point_inds, sampled_other_point_inds])

    data = data[sampled_point_inds]
    xyz = data[:, 0:3]
    rgb = data[:, 3:6]
    labels = data[:, 6].astype(int)

    xyz_min = np.amin(xyz, axis=0)
    xyz -= xyz_min
    if pc_augm:
        xyz = augment_pointcloud(xyz, pc_augm_config)
    if 'XYZ' in pc_attribs:
        xyz_min = np.amin(xyz, axis=0)
        XYZ = xyz - xyz_min
        xyz_max = np.amax(XYZ, axis=0)
        if np.any(xyz_max > 0):
            XYZ = XYZ / xyz_max
        else:
            XYZ = np.zeros_like(XYZ)

    ptcloud = []
    if 'xyz' in pc_attribs:
        ptcloud.append(xyz)
    if 'rgb' in pc_attribs:
        ptcloud.append(rgb / 255.)
    if 'XYZ' in pc_attribs:
        ptcloud.append(XYZ)
    ptcloud = np.concatenate(ptcloud, axis=1)

    if support:
        groundtruth = labels == sampled_class
    else:
        groundtruth = np.zeros_like(labels)
        for i, label in enumerate(labels):
            if label in sampled_classes:
                groundtruth[i] = sampled_classes.index(label) + 1

    return ptcloud, groundtruth


def augment_pointcloud(P, pc_augm_config):
    """" Augmentation on XYZ and jittering of everything """
    M = transforms3d.zooms.zfdir2mat(1)
    if pc_augm_config['scale'] > 1:
        s = random.uniform(1 / pc_augm_config['scale'], pc_augm_config['scale'])
        M = np.dot(transforms3d.zooms.zfdir2mat(s), M)
    if pc_augm_config['rot'] == 1:
        angle = random.uniform(0, 2 * math.pi)
        M = np.dot(transforms3d.axangles.axangle2mat([0, 0, 1], angle), M)  # z=upright assumption
    if pc_augm_config['mirror_prob'] > 0:  # mirroring x&y, not z
        if random.random() < pc_augm_config['mirror_prob'] / 2:
            M = np.dot(transforms3d.zooms.zfdir2mat(-1, [1, 0, 0]), M)
        if random.random() < pc_augm_config['mirror_prob'] / 2:
            M = np.dot(transforms3d.zooms.zfdir2mat(-1, [0, 1, 0]), M)
    P[:, :3] = np.dot(P[:, :3], M.T)

    if pc_augm_config['jitter']:
        sigma, clip = 0.01, 0.05
        P = P + np.clip(sigma * np.random.randn(*P.shape), -1 * clip, clip).astype(np.float32)
    return P


class MyDataset(Dataset):
    def __init__(self, data_path, dataset_name, cvfold=0, num_episode=50000, n_way=3, k_shot=5, n_queries=1,
                 phase=None, mode='train', num_point=2048, pc_attribs='xyzrgbXYZ', pc_augm=False, pc_augm_config=None):
        super(MyDataset, self).__init__()
        self.data_path = data_path
        self.n_way = n_way
        self.k_shot = k_shot
        self.n_queries = n_queries
        self.num_episode = num_episode
        self.phase = phase
        self.mode = mode
        self.num_point = num_point
        self.pc_attribs = pc_attribs
        self.pc_augm = pc_augm
        self.pc_augm_config = pc_augm_config

        if dataset_name == 's3dis':
            from dataloaders.s3dis import S3DISDataset
            self.dataset = S3DISDataset(cvfold, data_path)
        elif dataset_name == 'scannet':
            from dataloaders.scannet import ScanNetDataset
            self.dataset = ScanNetDataset(cvfold, data_path)
        else:
            raise NotImplementedError('Unknown dataset %s!' % dataset_name)

        if mode == 'train':
            self.classes = np.array(self.dataset.train_classes)
        elif mode == 'test':
            self.classes = np.array(self.dataset.test_classes)
        else:
            raise NotImplementedError('Unknown mode %s! [Options: train/test]' % mode)

        print('MODE: {0} | Classes: {1}'.format(mode, self.classes))
        self.class2scans = self.dataset.class2scans

    def __len__(self):
        return self.num_episode

    def __getitem__(self, index, n_way_classes=None):
        if n_way_classes is not None:
            sampled_classes = np.array(n_way_classes)
        else:
            sampled_classes = np.random.choice(self.classes, self.n_way, replace=False)

        support_ptclouds, support_masks, query_ptclouds, query_labels = self.generate_one_episode(sampled_classes)

        if self.mode == 'train' and self.phase == 'metatrain':
            remain_classes = list(set(self.classes) - set(sampled_classes))
            try:
                sampled_valid_classes = np.random.choice(np.array(remain_classes), self.n_way, replace=False)
            except:
                raise NotImplementedError('Error! The number remaining classes is less than %d_way' % self.n_way)

            valid_support_ptclouds, valid_support_masks, valid_query_ptclouds, \
            valid_query_labels = self.generate_one_episode(sampled_valid_classes)

            return (support_ptclouds.astype(np.float32),
                    support_masks.astype(np.int32),
                    query_ptclouds.astype(np.float32),
                    query_labels.astype(np.int64),
                    valid_support_ptclouds.astype(np.float32),
                    valid_support_masks.astype(np.int32),
                    valid_query_ptclouds.astype(np.float32),
                    valid_query_labels.astype(np.int64))
        else:
            # ✅ اصلاح: همیشه sampled_classes را در پایان برگردان (حتی در حالت تست/آموزش عادی)
            return (support_ptclouds.astype(np.float32),
                    support_masks.astype(np.int32),
                    query_ptclouds.astype(np.float32),
                    query_labels.astype(np.int64),
                    sampled_classes)  # ✅ این خط جدید است

    def generate_one_episode(self, sampled_classes):
        support_ptclouds = []
        support_masks = []
        query_ptclouds = []
        query_labels = []

        black_list = []  # to store the sampled scan names, in order to prevent sampling one scan several times...
        for sampled_class in sampled_classes:
            all_scannames = self.class2scans[sampled_class].copy()
            if len(black_list) != 0:
                all_scannames = [x for x in all_scannames if x not in black_list]
            if len(all_scannames) < self.k_shot + self.n_queries:
                # Fallback: allow replacement if not enough scans
                selected_scannames = np.random.choice(all_scannames, self.k_shot + self.n_queries, replace=True)
            else:
                selected_scannames = np.random.choice(all_scannames, self.k_shot + self.n_queries, replace=False)
            black_list.extend(selected_scannames)
            query_scannames = selected_scannames[:self.n_queries]
            support_scannames = selected_scannames[self.n_queries:]

            query_ptclouds_one_way, query_labels_one_way = sample_K_pointclouds(
                self.data_path, self.num_point, self.pc_attribs, self.pc_augm,
                self.pc_augm_config, query_scannames, sampled_class, sampled_classes, is_support=False)

            support_ptclouds_one_way, support_masks_one_way = sample_K_pointclouds(
                self.data_path, self.num_point, self.pc_attribs, self.pc_augm,
                self.pc_augm_config, support_scannames, sampled_class, sampled_classes, is_support=True)

            query_ptclouds.append(query_ptclouds_one_way)
            query_labels.append(query_labels_one_way)
            support_ptclouds.append(support_ptclouds_one_way)
            support_masks.append(support_masks_one_way)

        support_ptclouds = np.stack(support_ptclouds, axis=0)
        support_masks = np.stack(support_masks, axis=0)
        query_ptclouds = np.concatenate(query_ptclouds, axis=0)
        query_labels = np.concatenate(query_labels, axis=0)

        return support_ptclouds, support_masks, query_ptclouds, query_labels


def batch_train_task_collate(batch):
    task_train_support_ptclouds, task_train_support_masks, task_train_query_ptclouds, task_train_query_labels, \
    task_valid_support_ptclouds, task_valid_support_masks, task_valid_query_ptclouds, task_valid_query_labels = list(zip(*batch))

    task_train_support_ptclouds = np.stack(task_train_support_ptclouds)
    task_train_support_masks = np.stack(task_train_support_masks)
    task_train_query_ptclouds = np.stack(task_train_query_ptclouds)
    task_train_query_labels = np.stack(task_train_query_labels)
    task_valid_support_ptclouds = np.stack(task_valid_support_ptclouds)
    task_valid_support_masks = np.stack(task_valid_support_masks)
    task_valid_query_ptclouds = np.array(task_valid_query_ptclouds)
    task_valid_query_labels = np.stack(task_valid_query_labels)

    data = [torch.from_numpy(task_train_support_ptclouds).transpose(3, 4),
            torch.from_numpy(task_train_support_masks),
            torch.from_numpy(task_train_query_ptclouds).transpose(2, 3),
            torch.from_numpy(task_train_query_labels),
            torch.from_numpy(task_valid_support_ptclouds).transpose(3, 4),
            torch.from_numpy(task_valid_support_masks),
            torch.from_numpy(task_valid_query_ptclouds).transpose(2, 3),
            torch.from_numpy(task_valid_query_labels)]

    return data


################################################ Static Testing Dataset ################################################

class MyTestDataset(Dataset):
    def __init__(self, data_path, dataset_name, cvfold=0, num_episode_per_comb=100, n_way=3, k_shot=5, n_queries=1,
                 num_point=2048, pc_attribs='xyzrgbXYZ', mode='valid'):
        super(MyTestDataset, self).__init__()

        dataset = MyDataset(data_path, dataset_name, cvfold=cvfold, n_way=n_way, k_shot=k_shot, n_queries=n_queries,
                            mode='test', num_point=num_point, pc_attribs=pc_attribs, pc_augm=False)
        self.classes = dataset.classes

        # ✅ فیلتر کردن کلاس‌هایی که حداقل یک اسکن دارند
        filtered_classes = []
        for c in self.classes:
            if c in dataset.class2scans and len(dataset.class2scans[c]) > 0:
                filtered_classes.append(c)
            else:
                print(f"[Warning] Class {c} has no associated scans. Skipping.")
        self.classes = np.array(filtered_classes)

        if len(self.classes) < n_way:
            raise ValueError(f"Not enough valid test classes ({len(self.classes)}) for n_way={n_way}")

        if mode == 'valid':
            test_data_path = os.path.join(data_path, 'S_%d_N_%d_K_%d_episodes_%d_pts_%d' % (
                cvfold, n_way, k_shot, num_episode_per_comb, num_point))
        elif mode == 'test':
            test_data_path = os.path.join(data_path, 'S_%d_N_%d_K_%d_test_episodes_%d_pts_%d' % (
                cvfold, n_way, k_shot, num_episode_per_comb, num_point))
        else:
            raise NotImplementedError('Mode (%s) is unknown!' % mode)

        if os.path.exists(test_data_path):
            self.file_names = glob.glob(os.path.join(test_data_path, '*.h5'))
            self.num_episode = len(self.file_names)
        else:
            print('Test dataset (%s) does not exist...\n Constructing...' % test_data_path, flush=True)
            os.makedirs(test_data_path, exist_ok=True)

            class_comb = list(combinations(self.classes, n_way))
            self.num_episode = len(class_comb) * num_episode_per_comb

            episode_ind = 0
            self.file_names = []
            total_episodes = self.num_episode
            for idx_comb, sampled_classes in enumerate(class_comb):
                sampled_classes = list(sampled_classes)
                for i in range(num_episode_per_comb):
                    if episode_ind % 100 == 0:
                        print(f'  [Progress] Generating episode {episode_ind}/{total_episodes} ...', flush=True)
                    try:
                        data = dataset.__getitem__(episode_ind, sampled_classes)
                        out_filename = os.path.join(test_data_path, '%d.h5' % episode_ind)
                        write_episode(out_filename, data)
                        self.file_names.append(out_filename)
                    except Exception as e:
                        print(f"  [Warning] Skipping episode {episode_ind} due to error: {e}")
                        continue
                    episode_ind += 1
            print(f'  [Done] Generated {len(self.file_names)} test episodes.', flush=True)

    def __len__(self):
        return self.num_episode

    def __getitem__(self, index):
        file_name = self.file_names[index]
        return read_episode(file_name)


def batch_test_task_collate(batch):
    """
    ⚠️ This collate function is ONLY for training-time episodic data.
    For validation/test, use default collate (since data is pre-saved in H5).
    """
    support_ptclouds, support_masks, query_ptclouds, query_labels, _ = zip(*batch)

    # Stack supports: [B, n_way, k_shot, N, C] -> [B, n_way, k_shot, C, N]
    support_ptclouds = torch.from_numpy(np.stack(support_ptclouds)).transpose(3, 4)
    support_masks = torch.from_numpy(np.stack(support_masks))  # [B, n_way, k_shot, N]

    # Flatten ALL query points into a single list
    query_ptclouds = np.concatenate([q.reshape(-1, q.shape[-1]) for q in query_ptclouds], axis=0)  # [total_Q * N, C]
    query_labels = np.concatenate([l.reshape(-1) for l in query_labels], axis=0)                  # [total_Q * N]

    query_ptclouds = torch.from_numpy(query_ptclouds).transpose(0, 1)  # [C, total_Q * N]
    query_labels = torch.from_numpy(query_labels.astype(np.int64))     # [total_Q * N]

    return support_ptclouds, support_masks, query_ptclouds, query_labels


def write_episode(out_filename, data):
    # در حالت تست، data شامل 5 عنصر است
    support_ptclouds, support_masks, query_ptclouds, query_labels, sampled_classes = data
    data_file = h5.File(out_filename, 'w')
    data_file.create_dataset('support_ptclouds', data=support_ptclouds, dtype='float32')
    data_file.create_dataset('support_masks', data=support_masks, dtype='int32')
    data_file.create_dataset('query_ptclouds', data=query_ptclouds, dtype='float32')
    data_file.create_dataset('query_labels', data=query_labels, dtype='int64')
    data_file.create_dataset('sampled_classes', data=sampled_classes, dtype='int32')
    data_file.close()

    print('\t {0} saved! | classes: {1}'.format(out_filename, sampled_classes), flush=True)


def read_episode(file_name):
    data_file = h5.File(file_name, 'r')
    support_ptclouds = data_file['support_ptclouds'][:]
    support_masks = data_file['support_masks'][:]
    query_ptclouds = data_file['query_ptclouds'][:]
    query_labels = data_file['query_labels'][:]

    # ✅ اصلاح: پشتیبانی از فایل‌های قدیمی (بدون sampled_classes)
    if 'sampled_classes' in data_file:
        sampled_classes = data_file['sampled_classes'][:]
    else:
        # در صورت نبود، مقدار پیش‌فرض خالی
        sampled_classes = np.array([], dtype=np.int32)

    data_file.close()  # ✅ فایل را ببندید
    return support_ptclouds, support_masks, query_ptclouds, query_labels, sampled_classes


################################################  Pre-train Dataset ################################################
class MyPretrainDataset(Dataset):
    def __init__(self, data_path, classes, class2scans, mode='train', num_point=2048, pc_attribs='xyzrgbXYZ',
                 pc_augm=False, pc_augm_config=None):
        super(MyPretrainDataset, self).__init__()
        self.data_path = data_path
        self.classes = classes
        self.num_point = num_point
        self.pc_attribs = pc_attribs
        self.pc_augm = pc_augm
        self.pc_augm_config = pc_augm_config

        train_block_names = []
        all_block_names = []
        for k, v in sorted(class2scans.items()):
            all_block_names.extend(v)
            n_blocks = len(v)
            n_test_blocks = int(n_blocks * 0.1)
            n_train_blocks = n_blocks - n_test_blocks
            train_block_names.extend(v[:n_train_blocks])

        if mode == 'train':
            self.block_names = list(set(train_block_names))
        elif mode == 'test':
            self.block_names = list(set(all_block_names) - set(train_block_names))
        else:
            raise NotImplementedError('Mode is unknown!')

        print('[Pretrain Dataset] Mode: {0} | Num_blocks: {1}'.format(mode, len(self.block_names)))

    def __len__(self):
        return len(self.block_names)

    def __getitem__(self, index):
        block_name = self.block_names[index]

        ptcloud, label = sample_pointcloud(self.data_path, self.num_point, self.pc_attribs, self.pc_augm,
                                           self.pc_augm_config, block_name, self.classes, random_sample=True)

        return torch.from_numpy(ptcloud.transpose().astype(np.float32)), torch.from_numpy(label.astype(np.int64))


class MyFullDataset(Dataset):
    def __init__(self, data_path, dataset_name, cvfold=0, phase='train', classes=None,
                 num_point=2048, pc_attribs='xyzrgbXYZ', pc_augm=False, pc_augm_config=None):
        """
        Full (non-episodic) dataset for collecting base priors.
        Returns: (point_cloud [N, C], labels [N]) for scenes containing at least one point from `classes`.
        """
        super(MyFullDataset, self).__init__()
        self.data_path = data_path
        self.num_point = num_point
        self.pc_attribs = pc_attribs
        self.pc_augm = pc_augm
        self.pc_augm_config = pc_augm_config
        self.classes = classes if classes is not None else []

        if dataset_name == 's3dis':
            from dataloaders.s3dis import S3DISDataset
            self.dataset = S3DISDataset(cvfold, data_path)
        else:
            raise NotImplementedError(f"Dataset {dataset_name} not supported")

        # Determine relevant classes based on phase
        if phase == 'train':
            relevant_classes = self.dataset.train_classes
        else:
            relevant_classes = self.dataset.test_classes

        # If user specified `classes`, intersect with relevant_classes
        if self.classes:
            relevant_classes = [c for c in relevant_classes if c in self.classes]

        # Collect all scans that contain any of the relevant classes
        scene_set = set()
        for cls in relevant_classes:
            if cls in self.dataset.class2scans:
                scene_set.update(self.dataset.class2scans[cls])
        self.valid_scenes = list(scene_set)

        print(f"[MyFullDataset] Phase: {phase} | Classes: {relevant_classes} | Scenes: {len(self.valid_scenes)}")

    def __len__(self):
        return len(self.valid_scenes)

    def __getitem__(self, idx):
        scan_name = self.valid_scenes[idx]
        data = np.load(os.path.join(self.data_path, 'data', f'{scan_name}.npy'))
        xyz = data[:, 0:3]
        rgb = data[:, 3:6]
        labels = data[:, 6].astype(int)

        # ✅ فیلتر کردن فقط کلاس‌های مرتبط (برای base_priors تمیز)
        if self.classes:
            mask = np.isin(labels, self.classes)
            if not np.any(mask):
                # Fallback: return dummy if no relevant points
                xyz = np.zeros((self.num_point, 3), dtype=np.float32)
                rgb = np.zeros((self.num_point, 3), dtype=np.float32)
                labels = np.zeros(self.num_point, dtype=np.int64)
            else:
                xyz = xyz[mask]
                rgb = rgb[mask]
                labels = labels[mask]

        N = xyz.shape[0]
        if N == 0:
            xyz = np.zeros((self.num_point, 3), dtype=np.float32)
            rgb = np.zeros((self.num_point, 3), dtype=np.float32)
            labels = np.zeros(self.num_point, dtype=np.int64)
        elif N < self.num_point:
            pad_idx = np.random.choice(N, self.num_point - N, replace=True)
            xyz = np.concatenate([xyz, xyz[pad_idx]], axis=0)
            rgb = np.concatenate([rgb, rgb[pad_idx]], axis=0)
            labels = np.concatenate([labels, labels[pad_idx]], axis=0)
        else:
            choice = np.random.choice(N, self.num_point, replace=False)
            xyz = xyz[choice]
            rgb = rgb[choice]
            labels = labels[choice]

        # Normalize XYZ
        xyz_min = np.amin(xyz, axis=0)
        xyz_normalized = xyz - xyz_min
        xyz_max = np.amax(xyz_normalized, axis=0)
        if np.any(xyz_max > 0):
            xyz_normalized = xyz_normalized / xyz_max
        else:
            xyz_normalized = np.zeros_like(xyz_normalized)

        # Build point cloud
        ptcloud = []
        if 'xyz' in self.pc_attribs:
            ptcloud.append(xyz)
        if 'rgb' in self.pc_attribs:
            ptcloud.append(rgb / 255.0)
        if 'XYZ' in self.pc_attribs:
            ptcloud.append(xyz_normalized)
        ptcloud = np.concatenate(ptcloud, axis=1)

        # Apply augmentation (usually NOT for base_priors)
        if self.pc_augm and self.pc_augm_config is not None:
            ptcloud = augment_pointcloud(ptcloud, self.pc_augm_config)

        return torch.from_numpy(ptcloud).float(), torch.from_numpy(labels).long()
```

```python
%%writefile /content/attMPTI/scripts/train_attMPTI.sh
#!/usr/bin/env bash
set -euo pipefail

GPU_ID=0

DATASET='s3dis'
SPLIT=0
DATA_PATH='/content/attMPTI/datasets/S3DIS/blocks_bs1_s1'
SAVE_PATH='/content/log_s3dis'
LOG_DIR='/content/log_s3dis/mpti_run'

BATCH_SIZE=1
NUM_POINTS=2048
PC_ATTRIBS='xyzrgbXYZ'
EDGECONV_WIDTHS='[[64,64], [64,64], [64,64]]'
MLP_WIDTHS='[512,256]'
K=20
BASE_WIDTHS='[128,64]'

PRETRAIN_CHECKPOINT='/content/log_s3dis/log_pretrain_s3dis_S0'
N_WAY=2
K_SHOT=1          # ✅ تغییر اصلی: از 1 به 5
N_QUERIES=1
N_TEST_EPISODES=100

NUM_ITERS=42000
EVAL_INTERVAL=1000
LR=0.00005        # ✅ کاهش lr
DECAY_STEP=190
DECAY_RATIO=0.9998

N_SUBPROTOTYPES=20
K_CONNECT=50
SIGMA=1

USE_BAYESIAN=1
USE_ADAPTER=1
BETA=10.0         # ✅ افزایش beta
TEMPERATURE=0.8
PROJECTED_DIM=256 # ✅ افزایش بعد

TRAIN_STAGE=0
TRAIN_ALL_STAGES=0
USE_DEPTH_POLICY=0
POLICY_BINS=3
POLICY_SELECT_VIA='rule'
POLICY_DEPTH_FOR_GRAPH=1
POLICY_VAL_EPISODES=50

echo "NUM_ITERS=${NUM_ITERS}, EVAL_INTERVAL=${EVAL_INTERVAL}"

args=(--phase 'mptitrain'
      --dataset "${DATASET}" --cvfold "${SPLIT}"
      --data_path "${DATA_PATH}" --save_path "${SAVE_PATH}"
      --pretrain_checkpoint_path "${PRETRAIN_CHECKPOINT}"
      --use_attention
      --n_subprototypes "${N_SUBPROTOTYPES}" --k_connect "${K_CONNECT}"
      --sigma "${SIGMA}" --output_dim 64
      --pc_npts "${NUM_POINTS}" --pc_attribs "${PC_ATTRIBS}" --pc_augm
      --edgeconv_widths "${EDGECONV_WIDTHS}" --dgcnn_k "${K}"
      --dgcnn_mlp_widths "${MLP_WIDTHS}" --base_widths "${BASE_WIDTHS}"
      --n_iters "${NUM_ITERS}" --eval_interval "${EVAL_INTERVAL}" --batch_size "${BATCH_SIZE}"
      --lr "${LR}" --step_size "${DECAY_STEP}" --gamma "${DECAY_RATIO}"
      --n_way "${N_WAY}" --k_shot "${K_SHOT}" --n_queries "${N_QUERIES}" --n_episode_test "${N_TEST_EPISODES}"
      --projected_dim "${PROJECTED_DIM}"   # ⬅️ اضافه شده
)

if [[ "${USE_BAYESIAN}" == "1" ]]; then
  args+=(--use_bayesian --beta "${BETA}" --temperature "${TEMPERATURE}")
fi

if [[ "${USE_ADAPTER}" == "1" ]]; then
  args+=(--use_adapter)
fi

if [[ -n "${LOG_DIR}" ]]; then
  args+=(--log_dir "${LOG_DIR}")
fi

if [[ -n "${TRAIN_STAGE}" ]]; then
  args+=(--train_stage "${TRAIN_STAGE}")
fi

if [[ "${TRAIN_ALL_STAGES}" == "1" ]]; then
  args+=(--train_all_stages)
fi

# فقط اگر بیزین غیرفعال باشد
if [[ "${USE_DEPTH_POLICY}" == "1" ]]; then
  args+=(--use_depth_policy
         --policy_bins "${POLICY_BINS}"
         --policy_select_via "${POLICY_SELECT_VIA}"
         --policy_depth_for_graph "${POLICY_DEPTH_FOR_GRAPH}"
         --policy_val_episodes "${POLICY_VAL_EPISODES}")
fi

echo "Running with args:"
printf ' %q' "${args[@]}"; echo

CUDA_VISIBLE_DEVICES=$GPU_ID python /content/attMPTI/main.py "${args[@]}"
```

```python
%%writefile /content/attMPTI/runs/pre_train.py

""" Pre-train phase

Author: Zhao Na, 2020
"""

import os
import numpy as np

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch import optim
from torch.utils.data import DataLoader
from torch.utils.tensorboard import SummaryWriter

from dataloaders.loader import MyPretrainDataset
from models.dgcnn import DGCNN
from utils.logger import init_logger
from utils.checkpoint_util import save_pretrain_checkpoint


class DGCNNSeg(nn.Module):
    def __init__(self, args, num_classes):
        super(DGCNNSeg, self).__init__()
        self.encoder = DGCNN(args.edgeconv_widths, args.dgcnn_mlp_widths, args.pc_in_dim, k=args.dgcnn_k, return_edgeconvs=True)
        in_dim = args.dgcnn_mlp_widths[-1]
        for edgeconv_width in args.edgeconv_widths:
            in_dim += edgeconv_width[-1]
        self.segmenter = nn.Sequential(
                            nn.Conv1d(in_dim, 256, 1, bias=False),
                            nn.BatchNorm1d(256),
                            nn.LeakyReLU(0.2),
                            nn.Conv1d(256, 128, 1),
                            nn.BatchNorm1d(128),
                            nn.LeakyReLU(0.2),
                            nn.Dropout(0.3),
                            nn.Conv1d(128, num_classes, 1)
                         )

    def forward(self, pc):
        num_points = pc.shape[2]
        edgeconv_feats, point_feat = self.encoder(pc)
        global_feat = point_feat.max(dim=-1, keepdim=True)[0]
        edgeconv_feats.append(global_feat.expand(-1,-1,num_points))
        pc_feat = torch.cat(edgeconv_feats, dim=1)

        logits = self.segmenter(pc_feat)
        return logits


def metric_evaluate(predicted_label, gt_label, NUM_CLASS):
    """
    :param predicted_label: (B,N) tensor
    :param gt_label: (B,N) tensor
    :return: iou: scaler
    """
    gt_classes = [0 for _ in range(NUM_CLASS)]
    positive_classes = [0 for _ in range(NUM_CLASS)]
    true_positive_classes = [0 for _ in range(NUM_CLASS)]

    for i in range(gt_label.size()[0]):
        pred_pc = predicted_label[i]
        gt_pc = gt_label[i]

        for j in range(gt_pc.shape[0]):
            gt_l = int(gt_pc[j])
            pred_l = int(pred_pc[j])
            gt_classes[gt_l] += 1
            positive_classes[pred_l] += 1
            true_positive_classes[gt_l] += int(gt_l == pred_l)

    oa = sum(true_positive_classes)/float(sum(positive_classes))
    print('Overall accuracy: {0}'.format(oa))
    iou_list = []

    for i in range(NUM_CLASS):
        iou_class = true_positive_classes[i] / float(gt_classes[i]+positive_classes[i]-true_positive_classes[i])
        print('Class_%d: iou_class is %f' % (i, iou_class))
        iou_list.append(iou_class)

    mean_IoU = np.array(iou_list[1:]).mean()

    return oa, mean_IoU, iou_list


def pretrain(args):
    logger = init_logger(args.log_dir, args)

    # Init datasets, dataloaders, and writer
    PC_AUGMENT_CONFIG = {'scale': args.pc_augm_scale,
                         'rot': args.pc_augm_rot,
                         'mirror_prob': args.pc_augm_mirror_prob,
                         'jitter': args.pc_augm_jitter
                         }

    if args.dataset == 's3dis':
        from dataloaders.s3dis import S3DISDataset
        DATASET = S3DISDataset(args.cvfold, args.data_path)
    elif args.dataset == 'scannet':
        from dataloaders.scannet import ScanNetDataset
        DATASET = ScanNetDataset(args.cvfold, args.data_path)
    else:
        raise NotImplementedError('Unknown dataset %s!' % args.dataset)

    CLASSES = DATASET.train_classes
    NUM_CLASSES = len(CLASSES) + 1
    CLASS2SCANS = {c: DATASET.class2scans[c] for c in CLASSES}

    TRAIN_DATASET = MyPretrainDataset(args.data_path, CLASSES, CLASS2SCANS, mode='train',
                                      num_point=args.pc_npts, pc_attribs=args.pc_attribs,
                                      pc_augm=args.pc_augm, pc_augm_config=PC_AUGMENT_CONFIG)

    VALID_DATASET = MyPretrainDataset(args.data_path, CLASSES, CLASS2SCANS, mode='test',
                                      num_point=args.pc_npts, pc_attribs=args.pc_attribs,
                                      pc_augm=args.pc_augm, pc_augm_config=PC_AUGMENT_CONFIG)

    logger.cprint('=== Pre-train Dataset (classes: {0}) | Train: {1} blocks | Valid: {2} blocks ==='.format(
                                                     CLASSES, len(TRAIN_DATASET), len(VALID_DATASET)))

    TRAIN_LOADER = DataLoader(TRAIN_DATASET, batch_size=args.batch_size, num_workers=args.n_workers, shuffle=True,
                              drop_last=True)

    VALID_LOADER = DataLoader(VALID_DATASET, batch_size=args.batch_size, num_workers=args.n_workers, shuffle=False,
                              drop_last=True)

    WRITER = SummaryWriter(log_dir=args.log_dir)

    # Init model and optimizer
    model = DGCNNSeg(args, num_classes=NUM_CLASSES)
    print(model)
    if torch.cuda.is_available():
        model.cuda()

    optimizer = optim.Adam([{'params': model.encoder.parameters(), 'lr': args.pretrain_lr}, \
                           {'params': model.segmenter.parameters(), 'lr': args.pretrain_lr}], \
                            weight_decay=args.pretrain_weight_decay)
    # Set learning rate scheduler
    lr_scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=args.pretrain_step_size, gamma=args.pretrain_gamma)

    # train
    best_iou = 0
    global_iter = 0
    for epoch in range(args.pretrain_epochs):
        model.train()
        for batch_idx, (ptclouds, labels) in enumerate(TRAIN_LOADER):
            if torch.cuda.is_available():
                ptclouds = ptclouds.cuda()
                labels = labels.cuda()

            logits = model(ptclouds)
            loss = F.cross_entropy(logits, labels)

            # Loss backwards and optimizer updates
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

            WRITER.add_scalar('Train/loss', loss, global_iter)
            logger.cprint('=====[Train] Epoch: %d | Iter: %d | Loss: %.4f =====' % (epoch, batch_idx, loss.item()))
            global_iter += 1

        lr_scheduler.step()

        if (epoch+1) % args.eval_interval == 0:
            pred_total = []
            gt_total = []
            model.eval()
            with torch.no_grad():
                for i, (ptclouds, labels) in enumerate(VALID_LOADER):
                    gt_total.append(labels.detach())

                    if torch.cuda.is_available():
                        ptclouds = ptclouds.cuda()
                        labels = labels.cuda()

                    logits = model(ptclouds)
                    loss = F.cross_entropy(logits, labels)

                    # 　Compute predictions
                    _, preds = torch.max(logits.detach(), dim=1, keepdim=False)
                    pred_total.append(preds.cpu().detach())

                    WRITER.add_scalar('Valid/loss', loss, global_iter)
                    logger.cprint(
                        '=====[Valid] Epoch: %d | Iter: %d | Loss: %.4f =====' % (epoch, i, loss.item()))

            pred_total = torch.stack(pred_total, dim=0).view(-1, args.pc_npts)
            gt_total = torch.stack(gt_total, dim=0).view(-1, args.pc_npts)
            accuracy, mIoU, iou_perclass = metric_evaluate(pred_total, gt_total, NUM_CLASSES)
            logger.cprint('===== EPOCH [%d]: Accuracy: %f | mIoU: %f =====\n' % (epoch, accuracy, mIoU))
            WRITER.add_scalar('Valid/overall_accuracy', accuracy, global_iter)
            WRITER.add_scalar('Valid/meanIoU', mIoU, global_iter)

            if mIoU > best_iou:
                best_iou = mIoU
                logger.cprint('*******************Model Saved*******************')
                save_pretrain_checkpoint(model, args.log_dir)

    WRITER.close()
```
